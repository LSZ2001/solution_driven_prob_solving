```python
def maze_to_char_array(maze):
    return np.array([list(row) for row in maze])

def grid_to_rowstrings(s: str):
    grid = json.loads(s)          # -> list[list[str]]
    return [''.join(row) for row in grid]

def solve_maze_BestFirstSearch(maze, start=None, end=None, heuristic="Manhattan"):
    """
    Perform greedy best-first search on a rectangular maze.

    Parameters
    ----------
    maze : List[str]
        Maze represented as a list of strings of equal length.
        maze[row][col] == '#' for walls, '.' for free space.
    start : (col, row)
        Tuple giving the starting cell coordinates.
    end : (col, row)
        Tuple giving the end cell coordinates.
    heuristic: "Manhattan" or "Euclidean"
        Heuristic value function used to guide which square to evaluate next. 

    Returns
    -------
    path : List[(col, row)]
        The sequence of cells from start to end (inclusive), or [] if none found.
    evaluated : List[(col, row)]
        The cells popped from the priority queue, in order of expansion.
    walls : List[(col, row)]
        All wall‐cells that were examined as neighbors during the search.
    """
    width  = len(maze[0])
    height = len(maze)
    
    if(start is None):
        for x in range(width):
            if maze[0][x] == '.':
                start = (x, 0)
                break
        for x in range(width):
            if maze[height-1][x] == '.':
                end = (x, height-1)
                break

    def h(cell, distance="Manhattan"):
        c, r = cell
        gc, gr = end
        if(distance=="Manhattan"):
            return abs(c - gc) + abs(r - gr)
        elif(distance=="Euclidean"):
            return np.sqrt((c - gc)**2 + (r - gr)**2)

    # Min-heap of (priority, (col, row))
    frontier = [(h(start, distance=heuristic), start)]
    came_from = {start: None}
    visited   = {start}
    evaluated  = []
    walls      = []

    while frontier:
        _, current = heapq.heappop(frontier)
        evaluated.append(current)

        if current == end:
            break

        c, r = current
        for dc, dr in [(0,-1), (0,1), (-1,0), (1,0)]:
            nc, nr = c + dc, r + dr
            if not (0 <= nc < width and 0 <= nr < height):
                continue

            neighbor = (nc, nr)
            if neighbor in visited:
                continue

            if maze[nr][nc] == '#':
                walls.append(neighbor)
                continue

            visited.add(neighbor)
            came_from[neighbor] = current
            heapq.heappush(frontier, (h(neighbor, distance=heuristic), neighbor))

    # Reconstruct path
    path = []
    if end in came_from:
        node = end
        while node is not None:
            path.append(node)
            node = came_from[node]
        path.reverse()

    return path, evaluated, walls   
    

def visualize_maze_colored(maze,
                           num_colors,
                           figsize=None,
                           start_goal_pos = None,
                           start_goal=False,
                           path=None,
                           legend=False,
                           palette_name='tab20',
                           adj_sq_notsamecolor = True,
                           plt_show=True):
    """
    Visualize a maze with randomly assigned colors; but ensure no two
    adjacent cells of the same type share a color (greedy coloring).
    Parameters:
      - maze: list of strings ('#' walls, other symbols are all passages)
      - num_colors: total number of colors (must be >=2 and even)
      - figsize: tuple for figure size
      - start_goal: if True mark only start/end, else draw full path
      - path: list of (x,y) coordinates to overlay
      - legend: whether to draw the maze+legend (True) or just the maze (False)
      - palette_name: Matplotlib colormap name, e.g. 'tab20' or 'gray'
      - adj_sq_notsamecolor: if True, ensure no two adjacent cells of the same type share a color.
      - plt_show: if True, calls plt.show(); else leaves the figure open for saving downstream
    """
    # Randomize colors for each trial
    cmap = plt.get_cmap(palette_name)
    
    if(palette_name != "gray"):
        colors = [cmap(i) for i in range(20)]
        random.shuffle(colors)
        # 2. Split into wall-colors and passage-colors
        half = num_colors // 2
        wall_colors = colors[:half]
        passage_colors = colors[half:num_colors]
    else: # Black and white maze
#         positions = np.linspace(0, 1, num_colors)
        positions = np.linspace(0.85,1, num_colors)
        colors = [cmap(pos) for pos in positions]
        # 2. Split into wall-colors and passage-colors
        half = num_colors // 2
        wall_colors = colors[0:half]
        passage_colors = colors[-half:]

    H, W = len(maze), len(maze[0])
    # 2) Greedy coloring: no same‐type neighbors share a color
    img = np.zeros((H, W, 3), dtype=float)
    
    if(adj_sq_notsamecolor): # Ensure that no adjacent squares share the same color. Works only with num_colors>=8, due to the four-color theorem.
        assigned = [[None]*W for _ in range(H)]
        for y in range(H):
            for x in range(W):
                palette = wall_colors if maze[y][x] == '#' else passage_colors
                # collect forbidden from up and left if same type
                forbidden = set()
                for dy, dx in [(-1,0),(0,-1)]:
                    ny, nx = y+dy, x+dx
                    if 0 <= ny < H and 0 <= nx < W and maze[ny][nx] == maze[y][x]:
                        c = assigned[ny][nx]
                        if c is not None:
                            forbidden.add(c)
                # choose from palette avoiding forbidden
                choices = [c for c in palette if c not in forbidden]
                chosen = random.choice(choices) if choices else random.choice(palette)
                assigned[y][x] = chosen
                img[y, x] = chosen[:3]
    else:
        for y in range(H):
            for x in range(W):
                if maze[y][x] == '#':
                    color = random.choice(wall_colors)
                else:
                    color = random.choice(passage_colors)
                img[y, x] = color[:3]  # drop alpha channel

    # 3) Display
    if not legend:
        if figsize is None:
            plt.figure(figsize=(W/5, H/5))
        else:
            plt.figure(figsize=figsize)
        plt.imshow(img, interpolation='nearest')
        plt.xticks([]); plt.yticks([])
        if path:
            xs, ys = zip(*path)
            if(start_goal_pos is None):
                ax_maze.plot(xs[0], ys[0], '.', color='red', markersize=20)
                ax_maze.plot(xs[-1], ys[-1], '*', color='red', markersize=12)
            else:
                ax_maze.plot(start_goal_pos[0][0], start_goal_pos[0][1], '.', color='red', markersize=20)
                ax_maze.plot(start_goal_pos[1][0], start_goal_pos[1][1], '*', color='red', markersize=12)
            if not start_goal:
                plt.plot(xs, ys, color='red', linewidth=2)
        if plt_show:
            plt.show()
    else:
        # maze+legend side by side
        if figsize is None:
            figsize = ((W + half)/5, H/5)
        fig, (ax_maze, ax_leg) = plt.subplots(
            1, 2,
            figsize=figsize,
            gridspec_kw={'width_ratios': [W, max(8, num_colors)//2 * 1.2],
                         'wspace': 0.25}
        )
        # draw maze
        ax_maze.imshow(img, interpolation='nearest')
        ax_maze.set_xticks([]); ax_maze.set_yticks([])
        ax_maze.set_aspect('equal')
        if path:
            xs, ys = zip(*path)
            if(start_goal_pos is None):
                ax_maze.plot(xs[0], ys[0], '.', color='red', markersize=20)
                ax_maze.plot(xs[-1], ys[-1], '*', color='red', markersize=12)
            else:
                ax_maze.plot(start_goal_pos[0][0], start_goal_pos[0][1], '.', color='red', markersize=20)
                ax_maze.plot(start_goal_pos[1][0], start_goal_pos[1][1], '*', color='red', markersize=12)
            if not start_goal: # Visualize only start and end points of the maze, not path
                ax_maze.plot(xs, ys, color='red', linewidth=2)

        # draw legend
        legend_size = 1.5
        ax_leg.axis('off')
        ax_leg.set_aspect('equal')
        ax_leg.set_xlim(-1, (max(8, num_colors)//2 + 0.2)*legend_size)
        ax_leg.set_ylim(-2*legend_size, 3*legend_size)

        row_gap = 0.75
        # walls row
        for i, col in enumerate(wall_colors):
            ax_leg.add_patch(patches.Rectangle(
                (i*legend_size, legend_size + row_gap),
                legend_size, legend_size,
                facecolor=col[:3], edgecolor='k'
            ))
        ax_leg.text(-0.5, legend_size*1.5+row_gap, 'Danger:',
                    va='center', ha='right', fontsize=10)
        # spaces row
        for i, col in enumerate(passage_colors):
            ax_leg.add_patch(patches.Rectangle(
                (i*legend_size, 0),
                legend_size, legend_size,
                facecolor=col[:3], edgecolor='k'
            ))
        ax_leg.text(-0.5, legend_size*0.5, 'Safety:',
                    va='center', ha='right', fontsize=10)

        if plt_show:
            plt.show()
        else:
            fig.tight_layout()

    return {"ax_maze":ax_maze, "wall_colors": wall_colors,
            "passage_colors": passage_colors}

def path_on_rim(path, start_goal_pos = [(8,1),(8,15)], maze_size = 17):
    '''Given a path, check if it has never traversed the central maze.
        paths: 1D list. Each of its elements is one position along the path.
        start_goal_pos: starting and end point. Traversing these points does not mean the path is not on the rim.
        maze_size: used to determine where the rim positions are.'''
    paths_use_rim = []
    max_i = maze_size - 1
    if(path==[start_goal_pos[0]]): # If subject did not move at all, don't count this as using rim path
        return False

    for r, c in path:
        if (r, c) in set(start_goal_pos):
            continue
        # must be inside grid
        if not (0 <= r <= max_i and 0 <= c <= max_i):
            return False
        # on rim iff touches any outer edge
        if not (r in (0, max_i) or c in (0, max_i)):
            return False
    return True
        
    
def keypresses_to_position(key_presses, start_goal_pos=[(8,1),(8,15)], maze_size = 17):
    '''Given a set of key presses ['w','a','b',...], convert it to the corresponding trajectory of positions'''
    trajectory = [np.array(list(start_goal_pos[0]))] # start at start point
    
    for key_idx in range(len(key_presses)): #key_presses[0] moves from start point to first step. 
        prev_position = trajectory[key_idx]
        key = key_presses[key_idx]
        
        if(key=="b"): # Backtrack
            trajectory.append(trajectory[max(0,key_idx)])
            continue
        elif(key=="r"): # Return to start point
            trajectory.append(np.array(list(start_goal_pos[0]))) # Return to start
            continue
        else:
            if(key=="w"):
                move = np.array([0,-1])
            elif(key=="a"):
                move = np.array([-1,0])
            elif(key=="s"):
                move = np.array([0,1])
            elif(key=="d"):
                move = np.array([1,0])
            pseudo_pos = prev_position+move 
            pseudo_pos[0] = min(max(pseudo_pos[0],0), maze_size-1)
            pseudo_pos[1] = min(max(pseudo_pos[1],0), maze_size-1)
            trajectory.append(pseudo_pos)
        
    trajectory = [(int(pos[0]), int(pos[1])) for pos in trajectory]
    return trajectory
            
def find_last_occurrence_of_tuple(list_of_tuples, target_tuple):
    """
    Finds the index of the last occurrence of a target tuple in a list of tuples.

    Args:
        list_of_tuples: The list of tuples to search within.
        target_tuple: The tuple to find.

    Returns:
        The index of the last occurrence of the target tuple, or None if not found.
    """
    for r_idx, current_tuple in enumerate(reversed(list_of_tuples)):
        if current_tuple == target_tuple:
            return len(list_of_tuples) - 1 - r_idx
    return None        
 
    
def solve_maze_BestFirstSearch(maze, start=None, end=None, heuristic="Manhattan"):
    """
    Perform greedy best-first search on a rectangular maze.

    Parameters
    ----------
    maze : List[str]
        Maze represented as a list of strings of equal length.
        maze[row][col] == '#' for walls, '.' for free space.
    start : (col, row)
        Tuple giving the starting cell coordinates.
    end : (col, row)
        Tuple giving the end cell coordinates.
    heuristic: "Manhattan" or "Euclidean"
        Heuristic value function used to guide which square to evaluate next. 

    Returns
    -------
    path : List[(col, row)]
        The sequence of cells from start to end (inclusive), or [] if none found.
    evaluated : List[(col, row)]
        The cells popped from the priority queue, in order of expansion.
    walls : List[(col, row)]
        All wall‐cells that were examined as neighbors during the search.
    """
    width  = len(maze[0])
    height = len(maze)
    
    if(start is None):
        for x in range(width):
            if maze[0][x] == '.':
                start = (x, 0)
                break
        for x in range(width):
            if maze[height-1][x] == '.':
                end = (x, height-1)
                break

    def h(cell, distance="Manhattan"):
        c, r = cell
        gc, gr = end
        if(distance=="Manhattan"):
            return abs(c - gc) + abs(r - gr)
        elif(distance=="Euclidean"):
            return np.sqrt((c - gc)**2 + (r - gr)**2)

    # Min-heap of (priority, (col, row))
    frontier = [(h(start, distance=heuristic), start)]
    came_from = {start: None}
    visited   = {start}
    evaluated  = []
    walls      = []

    while frontier:
        _, current = heapq.heappop(frontier)
        evaluated.append(current)

        if current == end:
            break

        c, r = current
        for dc, dr in [(0,-1), (0,1), (-1,0), (1,0)]:
            nc, nr = c + dc, r + dr
            if not (0 <= nc < width and 0 <= nr < height):
                continue

            neighbor = (nc, nr)
            if neighbor in visited:
                continue

            if maze[nr][nc] == '#':
                walls.append(neighbor)
                continue

            visited.add(neighbor)
            came_from[neighbor] = current
            heapq.heappush(frontier, (h(neighbor, distance=heuristic), neighbor))

    # Reconstruct path
    path = []
    if end in came_from:
        node = end
        while node is not None:
            path.append(node)
            node = came_from[node]
        path.reverse()

    return path, evaluated, walls
   

# if __name__ == '__main__':
#     w = int(input("Enter maze width (cells): "))
#     h = int(input("Enter maze height (cells): "))
#     m = generate_maze(w, h)
#     p = solve_maze(m)
#     visualize_maze(m, p)


def make_vc_formula(group_col, random_slope_terms):
    '''This computes the random effects formula for BinomialBayesMixedGLM.
        Importantly, for categorical predictors, the random effects are separately specified for each value.'''
    vc = {f"ri_{group_col}": f"0 + C({group_col})"}  # random intercepts
    for term in random_slope_terms:
        key = f"rs_{group_col}_{term}".replace("(", "").replace(")", "").replace(" ", "")
        vc[key] = f"0 + C({group_col}):{term}"
    return vc

def logloss_brier_acc(y_true, p_pred, eps=1e-12):
    '''Given y_true (ground-truth labels) and p_pred from the fitted model, compute losses:
        nll: summed cross-entropy loss, divided by the number of trials.
        brier: MSE between y_true and p_pred.
        accuracy: using the 0.5 threshold over p_pred values, return binary prediction accuracy.
    '''
    p = np.clip(np.asarray(p_pred), eps, 1 - eps)
    y = np.asarray(y_true)
    nll = -np.sum(y * np.log(p) + (1 - y) * np.log(1 - p))
    return float(nll) / len(y), float(np.mean((p - y) ** 2)), float(np.mean((p >= 0.5).astype(int) == y))

def cv_glmm_kfold(df, response, group, fixed_formula, random_slopes, n_folds=5, random_state=0):
    '''Fit logistic GLME.
        df: dataset.
        response: y (binary)
        group: random effects are grouped by this variable (Subj)
        fixed_formula: fixed effects
        random_slopes: random effects; for categorical variables, separate random effects for each value they make take on.'''
    # sanity check
    uniq = set(np.unique(df[response]))
    if not uniq.issubset({0, 1}):
        raise ValueError(f"{response} must be 0/1, got {uniq}")

    formula = f"{response} ~ {fixed_formula}"
    vc = make_vc_formula(group, random_slopes)
    print(formula)
    print(vc)
    model_full = BinomialBayesMixedGLM.from_formula(formula, vc, df)
    res_full = model_full.fit_vb()

    kf = KFold(n_splits=n_folds, shuffle=True, random_state=random_state)

    rows = []
    for k, (tr_idx, te_idx) in enumerate(kf.split(df), start=1):
        train = df.iloc[tr_idx].copy()
        test  = df.iloc[te_idx].copy()

        # Fit model (Variational Bayes)
        model = BinomialBayesMixedGLM.from_formula(formula, vc, train)
        res = model.fit_vb()

        # ---- Build test fixed-effects design using training DesignInfo
        # res.model.data.design_info is the Patsy design for *fixed effects*
        design_info_fe = res.model.data.design_info   # fixed-effects DesignInfo
        # Construct test FE matrix consistent with the training coding/levels
        X_test = build_design_matrices([design_info_fe], test, return_type='dataframe')[0]

        # Extract fixed-effect coefficients (first q params)
        q = X_test.shape[1]
        beta = np.asarray(res.params[:q])

        # Predict probabilities via logistic link
        lin = X_test.values @ beta
        p_test = 1.0 / (1.0 + np.exp(-lin))

        # Losses
        mll, brier, acc = logloss_brier_acc(test[response].values, p_test)

        rows.append(dict(fold=k, n_train=len(train), n_test=len(test),
                         mean_logloss=mll, brier=brier, acc=acc))

    out = pd.DataFrame(rows)
    avg = out[["mean_logloss", "brier", "acc"]].mean()
    return out, avg.to_dict(), model_full, res_full


def cv_lmm_kfold(
    df, response, group, fixed_formula, random_formula,
    n_folds=5, random_state=0, include_re_on_test=True
):
    """
    K-fold CV for Linear Mixed Model (LMM).

    Parameters
    ----------
    df : DataFrame
    response : str              # e.g., "y"
    group : str                 # e.g., "Subj"
    fixed_formula : str         # e.g., "y ~ x1 + C(cond) + trial"
    random_formula : str        # e.g., "~ 1 + trial" or "~ 1"
    n_folds : int
    random_state : int
    include_re_on_test : bool   # if True, add Z*b_g for groups seen in training; unseen groups get RE=0

    Returns
    -------
    fold_table : DataFrame with per-fold MSE
    avg_mse : float
    model_full, res_full : fitted full-data model & results (REML=False)
    """
    # Fit once on full data for reference
    model_full = smf.mixedlm(fixed_formula, data=df, groups=df[group], re_formula=random_formula)
    res_full = model_full.fit(reml=False)

    kf = KFold(n_splits=n_folds, shuffle=True, random_state=random_state)
    rows = []

    for k, (tr_idx, te_idx) in enumerate(kf.split(df), start=1):
        train = df.iloc[tr_idx].copy()
        test  = df.iloc[te_idx].copy()

        # Fit on training fold
        model = smf.mixedlm(fixed_formula, data=train, groups=train[group], re_formula=random_formula)
        res = model.fit(reml=False)

        # ----- Fixed-effects test design (Patsy) using training design info
        fe_design_info = res.model.data.design_info
        X_test = build_design_matrices([fe_design_info], test, return_type='dataframe')[0]
        fe_beta = res.fe_params.to_numpy()
        y_pred = X_test.to_numpy() @ fe_beta  # population prediction

        if include_re_on_test:
            # ----- Random-effects design info from training, then build Z_test
            # Build training Z to capture its design_info (must mirror random_formula)
            Z_train = dmatrix(random_formula, train, return_type='dataframe')
            re_design_info = Z_train.design_info
            Z_test = build_design_matrices([re_design_info], test, return_type='dataframe')[0].to_numpy()

            # res.random_effects: dict mapping group label -> vector (length == n_random_effects)
            # Add Z*b_g for test rows whose group was seen in training
            re_by_group = res.random_effects  # {group_value: Series}
            test_groups = test[group].to_numpy()

            # Build an array of per-row RE contributions (default zeros)
            re_add = np.zeros(len(test_groups), dtype=float)

            # Pre-extract group->b_g vector (ndarray) once
            cache_bg = {g: np.asarray(b) for g, b in re_by_group.items()}

            # Efficient vectorized add: for each seen group in the test set, add Z_g @ b_g
            for g in np.unique(test_groups):
                if g in cache_bg:
                    idx = np.where(test_groups == g)[0]
                    b_g = cache_bg[g]
                    re_add[idx] = (Z_test[idx] @ b_g)

            y_pred = y_pred + re_add  # add random effects for seen groups

        # MSE on test fold
        y_true = test[response].to_numpy()
        mse = float(np.mean((y_true - y_pred) ** 2))

        rows.append(dict(fold=k, n_train=len(train), n_test=len(test), mse=mse))

    fold_table = pd.DataFrame(rows)
    avg_mse = float(fold_table["mse"].mean())
    return fold_table, avg_mse, model_full, res_full


def print_maze(maze):
    """Print the maze (list of strings) to stdout."""
    for row in maze:
        print(row)
        
def solve_maze_BestFirstSearch(maze, start=None, end=None, heuristic="Manhattan"):
    """
    Perform greedy best-first search on a rectangular maze.

    Parameters
    ----------
    maze : List[str]
        Maze represented as a list of strings of equal length.
        maze[row][col] == '#' for walls, '.' for free space.
    start : (col, row)
        Tuple giving the starting cell coordinates.
    end : (col, row)
        Tuple giving the end cell coordinates.
    heuristic: "Manhattan" or "Euclidean"
        Heuristic value function used to guide which square to evaluate next. 

    Returns
    -------
    path : List[(col, row)]
        The sequence of cells from start to end (inclusive), or [] if none found.
    evaluated : List[(col, row)]
        The cells popped from the priority queue, in order of expansion.
    walls : List[(col, row)]
        All wall‐cells that were examined as neighbors during the search.
    """
    width  = len(maze[0])
    height = len(maze)
    
    if(start is None):
        for x in range(width):
            if maze[0][x] == '.':
                start = (x, 0)
                break
        for x in range(width):
            if maze[height-1][x] == '.':
                end = (x, height-1)
                break

    def h(cell, distance="Manhattan"):
        c, r = cell
        gc, gr = end
        if(distance=="Manhattan"):
            return abs(c - gc) + abs(r - gr)
        elif(distance=="Euclidean"):
            return np.sqrt((c - gc)**2 + (r - gr)**2)

    # Min-heap of (priority, (col, row))
    frontier = [(h(start, distance=heuristic), start)]
    came_from = {start: None}
    visited   = {start}
    evaluated  = []
    walls      = []

    while frontier:
        _, current = heapq.heappop(frontier)
        evaluated.append(current)

        if current == end:
            break

        c, r = current
        for dc, dr in [(0,-1), (0,1), (-1,0), (1,0)]:
            nc, nr = c + dc, r + dr
            if not (0 <= nc < width and 0 <= nr < height):
                continue

            neighbor = (nc, nr)
            if neighbor in visited:
                continue

            if maze[nr][nc] == '#':
                walls.append(neighbor)
                continue

            visited.add(neighbor)
            came_from[neighbor] = current
            heapq.heappush(frontier, (h(neighbor, distance=heuristic), neighbor))

    # Reconstruct path
    path = []
    if end in came_from:
        node = end
        while node is not None:
            path.append(node)
            node = came_from[node]
        path.reverse()

    return path, evaluated, walls

def optimal_rim_length_difference(maze, start_goal_pos=[(8,1),(8,15)]):
    '''Computes the additional length traveled by using the rim path, compared to the optimal path through the maze.'''
    
    # 1) Find the optimal path length
    path, _, _ = solve_maze_BestFirstSearch(maze, start=start_goal_pos[0], end=start_goal_pos[1])
    optimal_path_length = len(path)-1
    
    # 2) Rim path length
    maze_H = len(maze)
    maze_W = len(maze[0])
    rim_path_length = (maze_H + maze_W)
    length_difference = rim_path_length - optimal_path_length
    
    # 3) Check if either rim path is open
    left_rim = [(0,y) for y in range(int((maze_W+1)/2)-1,-1,-1)]
    left_rim = left_rim + ([(x,0) for x in range(1,maze_H)])
    left_rim = left_rim + ([(maze_H-1,y) for y in range(1,int((maze_W+1)/2))])
    right_rim = [(0,y) for y in range(int((maze_W+1)/2)-1,maze_W)]
    right_rim = right_rim + ([(x,maze_W-1) for x in range(1,maze_H)])
    right_rim = right_rim + ([(maze_H-1,y) for y in range(maze_H-2, int((maze_W+1)/2)-2,-1)])
    maze_left_rim_open = np.prod(np.array([maze[pos[0]][pos[1]]!='#' for pos in left_rim]))
    maze_right_rim_open = np.prod(np.array([maze[pos[0]][pos[1]]!='#' for pos in right_rim]))
    rim_path_open = int(maze_left_rim_open==1 or maze_right_rim_open==1)
    
    return length_difference, optimal_path_length, rim_path_length, rim_path_open, maze_left_rim_open, maze_right_rim_open

def path_on_leftorright_rim(path, path_on_rim, maze):
    '''If the trial features a rim path, return whether it is the left or right rim path (0/1), or if both have been traversed (2).'''   
    if(path_on_rim!=1): # Subject did not only stay on the rims
        return -1
    if(len(path)==0): # No response for trial
        return -1
    
    maze_W = len(maze[0])

    cols_traversed = []
    for pos in path:
        if(pos[0]==((maze_W+1)/2-1)): # The central column (8) is not indicative
            continue;
        cols_traversed.append(pos[0])
    if(len(cols_traversed)==0): # Not complete path; only stayed in central column.
        return -1
    cols_traversed_unique = np.unique(np.array(cols_traversed))
    cols_traversed_onleft = np.prod(cols_traversed_unique < ((maze_W+1)/2-1)) # only left rim is traversed
    cols_traversed_onright = np.prod(cols_traversed_unique > ((maze_W+1)/2-1)) # only right rim is traversed
    if(cols_traversed_onleft+cols_traversed_onright==2): # Both rim paths were traversed
        return 2
    elif(cols_traversed_onleft==1  and cols_traversed_onright==0): # Used left path only
        return 0
    elif(cols_traversed_onleft==0  and cols_traversed_onright==1): # Used right path only
        return 1

    
def num_squares_to_inspect(pos, maze_H, maze_W):
    '''Given a position tuple in a maze, 
    return (the number of adjacent squares - 1) because the previous pos need not be inspected again.
    So for corner pos -> return 1
           edge pos -> return 2
           other pos -> return 3'''   
    num_squares_adjacent = 3
    if(pos[1]==(maze_H-1) or pos[1]==0):
        num_squares_adjacent = num_squares_adjacent - 1
    if(pos[0]==(maze_W-1) or pos[0]==0):
        num_squares_adjacent = num_squares_adjacent - 1
        
    return num_squares_adjacent

def sum_num_squares_to_inspect(path, maze):
    '''Return #squares to inspect along the entire path, counting backtracks.
    '''
    maze_H = len(maze)
    maze_W = len(maze[0])
    
    num_squares_adjacents = []
    for pos in path[0:-1]: # Do not count the last pos, which is the goal pos---no need to inspect any squares there.
        num_squares_adjacents.append(num_squares_to_inspect(pos, maze_H, maze_W))
    num_squares_adjacents = np.array(num_squares_adjacents)
    return np.sum(num_squares_adjacents)
```

## II) Policy compression as an explanation?

I want to check when subjects use rim paths, do they have knowledge of which rim path is open (if only one path is open)? Or are they just randomly choosing one of the two rim paths (corresponding to beta=0; subjects ignore the color legend and don't verify their submitted path)?

Similarly, when no rim path is open, do subjects submit rim paths less?

```python
# Good news: answer is no. 

df_temp = df_allsubjs_test[df_allsubjs_test["num_colors"]==8] # Only the num_colors = 8 condition?
#df_temp = df_allsubjs_test.copy()

df_temp = df_temp[df_temp["path_on_rim"]==1] # Trials where participant used rim paths
df_temp = df_temp[df_temp["left_rim_path_open"] != df_temp["right_rim_path_open"]] # Trials where only one out of the two rim paths is open
df_temp = df_temp[~np.isnan(df_temp["path_taken_on_leftorrightrim"])]
df_temp["sensitive_to_whichrimopen"] = (df_temp["right_rim_path_open"]==df_temp["path_taken_on_leftorrightrim"]).astype(int) # For the above trials, did the subjects submit the (unique) open rim path?

# Confusion matrix (less good than the below, because it does not consider mean for each subject)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(df_temp["right_rim_path_open"], df_temp["path_taken_on_leftorrightrim"])
print(cm)
print(scipy.stats.chi2_contingency(cm))
print()

# Take mean over subjects
df_temp_avg = df_temp.groupby(['Subj'], as_index=False)[["Subj","cond","first_cond","trajectory_length",
       "hit_wall", "reached_goal", "submitted_in_time", "trial_success",
       "trial_bonus","path_on_rim","full_path_on_rim", "rimpath_agreement", "rt", "first_rt", "submitted_path_rt", "submitted_path_presses", "submitted_path_avgrt","right_rim_path_open","path_taken_on_leftorrightrim","sensitive_to_whichrimopen"]].mean()
print("P(choose correct rim path): {0} +- {1}".format(df_temp_avg["sensitive_to_whichrimopen"].mean(), df_temp_avg["sensitive_to_whichrimopen"].std(ddof=1)/np.sqrt(len(df_temp_avg))))
print(df_temp_avg[["Subj","sensitive_to_whichrimopen"]])

# Check if correct choice prob is statisticaly significantly above chance (0.5)
print("t-test 1samp:", scipy.stats.ttest_1samp(df_temp_avg["sensitive_to_whichrimopen"], 0.5))
print("Wilcoxon test:", scipy.stats.wilcoxon(df_temp_avg["sensitive_to_whichrimopen"]-0.5))
```

```python
# Good news: the above effects are not merely due to seeing the easy cond first.
# No matter which cond subjects did first, they are always sensitive to "which rim is open"!
for first_cond in [0,1]:
    df_temp_avg_temp = df_temp_avg[df_temp_avg["first_cond"]==first_cond]
    print("P(choose correct rim path): {0} +- {1}".format(df_temp_avg_temp["sensitive_to_whichrimopen"].mean(), df_temp_avg_temp["sensitive_to_whichrimopen"].std(ddof=1)/np.sqrt(len(df_temp_avg_temp))))

    # Check if correct choice prob is statisticaly significantly above chance (0.5)
    print("t-test 1samp:", scipy.stats.ttest_1samp(df_temp_avg_temp["sensitive_to_whichrimopen"], 0.5))
    print("Wilcoxon test:", scipy.stats.wilcoxon(df_temp_avg_temp["sensitive_to_whichrimopen"]-0.5))
    print()
```

# 1) Figures

```python
def annotate_paired(
    ax, df, x, y, hue=None, pairs=None,
    order=None, hue_order=None,
    cat_width=0.8,       # match seaborn's violinplot width
    text_offset=0.06,    # fraction of y-range between brackets
    linewidth=1.2, fontsize=12,
    label='stars',       # 'stars' | 'p' | 'p_and_d'
    test_type = "paired-t", # "paired-t" | "wilcoxon"
    print_table=True,    # print a table of results
    return_table=False,  # return the results DataFrame
    decimals=3           # rounding in printed table
):
    """
    Draw paired t-test brackets and (optionally) print/return stats.
    If hue is None: pairs = [(x1, x2), ...]
    If hue is not None: pairs = [((x1,h1),(x2,h2)), ...]

    Effect size reported: Cohen's dz = mean(diff)/sd(diff), paired design.
    """
    # Ensure Subject column for pairing
    if 'Subj' not in df.columns:
        df = df.reset_index()

    # Resolve plotting orders
    if order is None:
        order = list(pd.unique(df[x]))
    if hue is not None and hue_order is None:
        hue_order = list(pd.unique(df[hue]))

    # ---- centers that match split-violins exactly ----
    if hue is None:
        offset_map = {None: 0.0}
    else:
        k = len(hue_order)
        cell = cat_width / float(k)  # width per split violin
        centers = np.linspace(-cat_width/2 + cell/2,
                              +cat_width/2 - cell/2, k)
        offset_map = {h: centers[i] for i, h in enumerate(hue_order)}

    def xpos(x_level, h_level=None):
        xi = order.index(x_level)
        return xi + (offset_map[None] if hue is None else offset_map[h_level])

    # ---- compute stats for each pair ----
    rows = []
    for pr in pairs:
        if hue is None:
            x1, x2 = pr
            d1 = df[df[x] == x1][['Subj', y]].rename(columns={y:'v1'})
            d2 = df[df[x] == x2][['Subj', y]].rename(columns={y:'v2'})
            lab = f"{x} {x1} vs {x2}"
        else:
            (x1, h1), (x2, h2) = pr
            d1 = df[(df[x] == x1) & (df[hue] == h1)][['Subj', y]].rename(columns={y:'v1'})
            d2 = df[(df[x] == x2) & (df[hue] == h2)][['Subj', y]].rename(columns={y:'v2'})
            lab = f"({x1},{h1}) vs ({x2},{h2})"

        merged = pd.merge(d1, d2, on='Subj').dropna(subset=['v1','v2'])
        n = len(merged)
        if n >= 2:
            diff = (merged['v1'] - merged['v2']).to_numpy(dtype=float)
            sd = np.std(diff, ddof=1)
            dz = np.nan if sd == 0 else np.mean(diff) / sd
            if(test_type=="paired-t"):
                t_stat, p = ttest_rel(merged['v1'], merged['v2'])
            elif(test_type=="wilcoxon"):
                t_stat, p = wilcoxon(merged['v1'], merged['v2'])
            dfree = n - 1
        else:
            t_stat = p = dz = np.nan
            dfree = n - 1

        rows.append({
            'comparison': lab, 'n': n, 'df': dfree,
            't': t_stat, 'p': p, 'cohen_dz': dz
        })

    results = pd.DataFrame(rows)

    # ---- vertical layout & draw ----
    ymin, ymax = ax.get_ylim()
    yr = max(ymax - ymin, 1e-9)
    step = text_offset * yr
    ax.set_ylim(ymin, ymax + step * (len(pairs)*3 + 1))

    for i, pr in enumerate(pairs):
        if hue is None:
            x1, x2 = pr
            x1p, x2p = xpos(x1), xpos(x2)
        else:
            (x1, h1), (x2, h2) = pr
            x1p, x2p = xpos(x1, h1), xpos(x2, h2)

        stats = results.iloc[i]
        p = stats['p']; dz = stats['cohen_dz']

        # label text
        if label == 'stars':
            if   np.isnan(p): s = 'n/a'
            elif p < 1e-3:    s = '***'
            elif p < 1e-2:    s = '**'
            elif p < 5e-2:    s = '*'
            else:             s = 'n.s.'
        elif label == 'p':
            s = f"p={p:.3g}" if not np.isnan(p) else "p=n/a"
        elif label == 'p_and_d':
            sp = f"p={p:.3g}" if not np.isnan(p) else "p=n/a"
            sd = f"dᶻ={dz:.2f}" if not np.isnan(dz) else "dᶻ=n/a"
            s = sp + "\n" + sd
        else:
            s = ''

        # bracket height for this pair
        yb = ymax + step * (i*3 + 1)
        ytop = yb + step

        ax.plot([x1p, x1p, x2p, x2p], [yb, ytop, ytop, yb],
                'k-', lw=linewidth, clip_on=False)
        ax.text((x1p+x2p)/2.0, ytop + 0.12*step, s,
                ha='center', va='bottom', fontsize=fontsize, clip_on=False)

    # ---- print & return ----
    if print_table:
        with pd.option_context('display.float_format', lambda v: f"{v:.{decimals}f}"):
            print(results.to_string(index=False))

    if return_table:
        return results

        
```

```python
magnify = 2.5
condition_names = ["Easy", "Hard"]
colorbar_mode = "shared"   # "shared" or "each"

# dataframe column names
subj_col = "Subj"
maze_col = "test_maze_orig_idx"
trajectory_col = "trajectory_parsed"
condition_col = "cond"
easy_value = 0
hard_value = 1

# set depending on your coordinate convention
swap_xy = True

# if True: each subject contributes at most 1 per visited cell
# if False: repeated visits count multiple times
binary_per_subject = False


# =========================================================
# Helpers
# =========================================================
def maze_to_char_array(maze):
    return np.array([list(row) for row in maze])


def occupancy_from_path_list(path_list, maze_shape, swap_xy=False, binary_per_subject=False):
    occ = np.zeros(maze_shape, dtype=float)

    for path in path_list:
        subj_occ = np.zeros(maze_shape, dtype=float)

        for pos in path:
            x, y = pos[:2]
            x = int(x)
            y = int(y)

            if swap_xy:
                rr, cc = y, x
            else:
                rr, cc = x, y

            if 0 <= rr < maze_shape[0] and 0 <= cc < maze_shape[1]:
                if binary_per_subject:
                    subj_occ[rr, cc] = 1
                else:
                    subj_occ[rr, cc] += 1

        occ += subj_occ

    return occ


def make_maze_background(maze_arr):
    """
    Top-row maze rendering only.
    wall(#) -> black
    everything else -> white
    """
    bg = np.ones(maze_arr.shape, dtype=float)
    bg[maze_arr == '#'] = 0.0
    return bg


def draw_maze_layout(ax, maze_arr, title=None):
    bg = make_maze_background(maze_arr)
    ax.imshow(bg, cmap="gray", origin="upper", interpolation="none")

    ax.set_xticks(np.arange(-0.5, maze_arr.shape[1], 1), minor=True)
    ax.set_yticks(np.arange(-0.5, maze_arr.shape[0], 1), minor=True)
    ax.grid(which="minor", color="lightgray", linestyle="-", linewidth=0.4)

    ax.tick_params(
        which='both',
        bottom=False, left=False,
        labelbottom=False, labelleft=False
    )

    if title is not None:
        ax.set_title(title)


def draw_heatmap(ax, maze_arr, occ, ylabel=None, vmin=None, vmax=None):
    """
    Pure heatmap: no maze masking / no wall blackout.
    """
    hm = ax.imshow(
        occ,
        cmap="hot",
        origin="upper",
        interpolation="nearest",
        vmin=vmin,
        vmax=vmax
    )

    ax.set_xticks(np.arange(-0.5, maze_arr.shape[1], 1), minor=True)
    ax.set_yticks(np.arange(-0.5, maze_arr.shape[0], 1), minor=True)
    ax.grid(which="minor", color="lightgray", linestyle="-", linewidth=0.4)

    ax.tick_params(
        which='both',
        bottom=False, left=False,
        labelbottom=False, labelleft=False
    )

    if ylabel is not None:
        ax.set_ylabel(ylabel, fontsize=12)

    return hm


# =========================================================
# Build pooled occupancy maps from df_allsubjs_test
# =========================================================
df_plot = df_allsubjs_test.copy()

required_cols = [subj_col, maze_col, trajectory_col, condition_col]
missing = [c for c in required_cols if c not in df_plot.columns]
if len(missing) > 0:
    raise ValueError(f"Missing required columns: {missing}")

df_plot = df_plot[required_cols].copy()
df_plot = df_plot[df_plot[condition_col].isin([easy_value, hard_value])].copy()
df_plot = df_plot[df_plot[trajectory_col].notna()].copy()

maze_ids = sorted(df_plot[maze_col].dropna().unique())
n_maze = len(maze_ids)

maze_arrays = []
occupancies_easy = []
occupancies_hard = []

for maze_idx in maze_ids:
    maze_arr = maze_to_char_array(test_mazes[int(maze_idx)])
    maze_shape = maze_arr.shape

    df_maze = df_plot[df_plot[maze_col] == maze_idx]

    paths_easy = df_maze.loc[df_maze[condition_col] == easy_value, trajectory_col].tolist()
    paths_hard = df_maze.loc[df_maze[condition_col] == hard_value, trajectory_col].tolist()

    occ_easy = occupancy_from_path_list(
        paths_easy, maze_shape,
        swap_xy=swap_xy,
        binary_per_subject=binary_per_subject
    )
    occ_hard = occupancy_from_path_list(
        paths_hard, maze_shape,
        swap_xy=swap_xy,
        binary_per_subject=binary_per_subject
    )

    maze_arrays.append(maze_arr)
    occupancies_easy.append(occ_easy)
    occupancies_hard.append(occ_hard)

# shared scale if requested
global_vmax = max(
    max(np.max(x) for x in occupancies_easy) if len(occupancies_easy) > 0 else 0,
    max(np.max(x) for x in occupancies_hard) if len(occupancies_hard) > 0 else 0
)


# =========================================================
# Plot: top row maze, second easy, third hard
# =========================================================
n_rows_total = 3

fig, ax = plt.subplots(
    n_rows_total,
    n_maze,
    figsize=(n_maze * magnify, n_rows_total * magnify * 1.4),
    gridspec_kw={'wspace': 0.05, 'hspace': 0.10},
    constrained_layout=True
)

if n_maze == 1:
    ax = np.array(ax).reshape(n_rows_total, 1)

# -------------------------
# Top row: actual maze layouts
# -------------------------
for col_idx, maze_idx in enumerate(maze_ids):
    draw_maze_layout(
        ax[0, col_idx],
        maze_arrays[col_idx],
        title=f"Maze {maze_idx}"
    )
    if col_idx == 0:
        ax[0, col_idx].set_ylabel("Maze", fontsize=12)

# -------------------------
# Bottom rows: pooled heatmaps
# -------------------------
all_hm = []

for col_idx, maze_idx in enumerate(maze_ids):
    maze_arr = maze_arrays[col_idx]
    occs = [occupancies_easy[col_idx], occupancies_hard[col_idx]]

    for cond_idx, occ in enumerate(occs):
        cur_ax = ax[1 + cond_idx, col_idx]

        if colorbar_mode == "shared":
            hm = draw_heatmap(
                cur_ax,
                maze_arr,
                occ,
                ylabel=condition_names[cond_idx] if col_idx == 0 else None,
                vmin=0,
                vmax=global_vmax
            )
        elif colorbar_mode == "each":
            local_vmax = np.max(occ)
            hm = draw_heatmap(
                cur_ax,
                maze_arr,
                occ,
                ylabel=condition_names[cond_idx] if col_idx == 0 else None,
                vmin=0,
                vmax=local_vmax if local_vmax > 0 else 1
            )
            cbar = fig.colorbar(hm, ax=cur_ax, shrink=0.85, pad=0.01)
            if binary_per_subject:
                cbar.set_label("N subjects")
            else:
                cbar.set_label("Visit count")
        else:
            raise ValueError("colorbar_mode must be 'shared' or 'each'")

        all_hm.append(hm)

if colorbar_mode == "shared":
    cbar = fig.colorbar(all_hm[-1], ax=ax[1:, :], shrink=0.9, pad=0.01)
    if binary_per_subject:
        cbar.set_label("State occupancy (N subjects)")
    else:
        cbar.set_label("State occupancy (visit count)")

plt.suptitle("Maze layouts and pooled occupancy heatmaps from df_allsubjs_test", y=1.02)

fname = f"exp1_occupancy_heatmaps_from_df_colorbar-{colorbar_mode}.png"
plt.savefig(os.path.join(figpath, fname), dpi=300, bbox_inches="tight")
plt.show()
```

```python
# =========================================================
# Helpers
# =========================================================
def normalize_occ(occ, eps=0.0):
    x = occ.astype(float).copy()
    if eps > 0:
        x = x + eps
    s = x.sum()
    if s <= 0:
        return np.full_like(x, 1.0 / x.size)
    return x / s


def occupancy_from_single_path(path, maze_shape, swap_xy=False, binary=True):
    occ = np.zeros(maze_shape, dtype=float)

    for pos in path:
        x, y = pos[:2]
        x = int(x)
        y = int(y)

        if swap_xy:
            rr, cc = y, x
        else:
            rr, cc = x, y

        if 0 <= rr < maze_shape[0] and 0 <= cc < maze_shape[1]:
            if binary:
                occ[rr, cc] = 1
            else:
                occ[rr, cc] += 1

    return occ


def occupancy_from_path_list(path_list, maze_shape, swap_xy=False, binary_per_subject=True):
    occ = np.zeros(maze_shape, dtype=float)

    for path in path_list:
        subj_occ = occupancy_from_single_path(
            path, maze_shape,
            swap_xy=swap_xy,
            binary=binary_per_subject
        )
        occ += subj_occ

    return occ


# =========================================================
# Metrics
# =========================================================
def js_divergence_occupancy(occ1, occ2):
    p = normalize_occ(occ1).ravel()
    q = normalize_occ(occ2).ravel()
    return float(jensenshannon(p, q, base=2) ** 2)


def wasserstein_2d_exact(occ1, occ2, metric="euclidean"):
    p = normalize_occ(occ1)
    q = normalize_occ(occ2)

    coords1 = np.argwhere(p > 0)
    coords2 = np.argwhere(q > 0)

    w1 = p[p > 0]
    w2 = q[q > 0]

    if len(coords1) == 0 or len(coords2) == 0:
        return np.nan

    C = cdist(coords1, coords2, metric=metric)
    m, n = C.shape

    c = C.ravel()

    A_eq = np.zeros((m + n, m * n), dtype=float)
    b_eq = np.concatenate([w1, w2])

    for i in range(m):
        A_eq[i, i * n:(i + 1) * n] = 1.0

    for j in range(n):
        A_eq[m + j, j::n] = 1.0

    bounds = [(0, None)] * (m * n)

    res = linprog(c=c, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method="highs")

    if not res.success:
        raise RuntimeError(f"Wasserstein LP failed: {res.message}")

    return float(res.fun)


def rim_change_single(easy_rim, hard_rim):
    return float(hard_rim - easy_rim)


# =========================================================
# Main function using df_allsubjs_test
# =========================================================
def compute_trace_change_metrics_from_df(
    df_allsubjs_test,
    test_mazes,
    condition_col,
    easy_value,
    hard_value,
    subj_col="Subj",
    maze_col="test_maze_orig_idx",
    trajectory_col="trajectory_parsed",
    rim_col="path_on_rim",
    swap_xy=False,
    binary_per_path=True,
    wasserstein_metric="euclidean",
):
    """
    Compute subject/maze/overall trace-change metrics from df_allsubjs_test.

    Required columns in df_allsubjs_test:
        subj_col
        maze_col
        condition_col
        trajectory_col
        rim_col

    Parameters
    ----------
    condition_col : str
        Name of the column identifying easy vs hard condition.
    easy_value, hard_value :
        The values in condition_col corresponding to easy and hard.

    Returns
    -------
    df_subj_maze, df_subject, df_maze, df_overall
    """

    df = df_allsubjs_test.copy()

    required_cols = [subj_col, maze_col, condition_col, trajectory_col, rim_col]
    missing = [c for c in required_cols if c not in df.columns]
    if len(missing) > 0:
        raise ValueError(f"Missing required columns: {missing}")

    # keep only needed columns, and only easy/hard rows
    df = df[required_cols].copy()
    df = df[df[condition_col].isin([easy_value, hard_value])].copy()

    # drop rows with missing trajectories
    df = df[df[trajectory_col].notna()].copy()

    # Make sure trajectory entries are usable
    # (assumes they are already parsed lists of tuples, as you indicated)
    # If needed, you can add literal_eval here, but I am not doing it by default.

    # ---------------------------------
    # 1) Subject x Maze table
    # ---------------------------------
    rows_subj_maze = []

    grouped = df.groupby([subj_col, maze_col], sort=True)

    for (subj, maze_idx), g in grouped:
        g_easy = g[g[condition_col] == easy_value]
        g_hard = g[g[condition_col] == hard_value]

        if len(g_easy) == 0 or len(g_hard) == 0:
            continue

        # if duplicates exist, just take first
        row_easy = g_easy.iloc[0]
        row_hard = g_hard.iloc[0]

        maze_arr = maze_to_char_array(test_mazes[int(maze_idx)])
        maze_shape = maze_arr.shape

        path_easy = row_easy[trajectory_col]
        path_hard = row_hard[trajectory_col]

        occ_easy = occupancy_from_single_path(
            path_easy, maze_shape,
            swap_xy=swap_xy,
            binary=binary_per_path
        )
        occ_hard = occupancy_from_single_path(
            path_hard, maze_shape,
            swap_xy=swap_xy,
            binary=binary_per_path
        )

        w2d = wasserstein_2d_exact(occ_easy, occ_hard, metric=wasserstein_metric)
        jsd = js_divergence_occupancy(occ_easy, occ_hard)

        easy_rim = row_easy[rim_col]
        hard_rim = row_hard[rim_col]
        rim_delta = rim_change_single(easy_rim, hard_rim)

        rows_subj_maze.append({
            "subject": subj,
            "maze": maze_idx,
            "wasserstein_2d": w2d,
            "js_divergence": jsd,
            "rim_change": rim_delta,
            "easy_rim": easy_rim,
            "hard_rim": hard_rim,
            "easy_path_len": len(path_easy),
            "hard_path_len": len(path_hard),
        })

    df_subj_maze = pd.DataFrame(rows_subj_maze)

    # ---------------------------------
    # 2) Subject-level summary
    # ---------------------------------
    df_subject = (
        df_subj_maze
        .groupby("subject", as_index=False)
        .agg({
            "wasserstein_2d": "mean",
            "js_divergence": "mean",
            "rim_change": "mean",
            "easy_rim": "mean",
            "hard_rim": "mean",
            "easy_path_len": "mean",
            "hard_path_len": "mean",
        })
        .rename(columns={
            "easy_rim": "prop_easy_rim",
            "hard_rim": "prop_hard_rim",
            "easy_path_len": "mean_easy_path_len",
            "hard_path_len": "mean_hard_path_len",
        })
    )

    # ---------------------------------
    # 3) Maze-level pooled summary
    # ---------------------------------
    rows_maze = []

    for maze_idx, g_maze in df.groupby(maze_col, sort=True):
        g_easy = g_maze[g_maze[condition_col] == easy_value]
        g_hard = g_maze[g_maze[condition_col] == hard_value]

        if len(g_easy) == 0 or len(g_hard) == 0:
            continue

        maze_arr = maze_to_char_array(test_mazes[int(maze_idx)])
        maze_shape = maze_arr.shape

        paths_easy = g_easy[trajectory_col].tolist()
        paths_hard = g_hard[trajectory_col].tolist()

        occ_easy_pool = occupancy_from_path_list(
            paths_easy, maze_shape,
            swap_xy=swap_xy,
            binary_per_subject=binary_per_path
        )
        occ_hard_pool = occupancy_from_path_list(
            paths_hard, maze_shape,
            swap_xy=swap_xy,
            binary_per_subject=binary_per_path
        )

        w2d = wasserstein_2d_exact(occ_easy_pool, occ_hard_pool, metric=wasserstein_metric)
        jsd = js_divergence_occupancy(occ_easy_pool, occ_hard_pool)

        easy_rim_mean = g_easy[rim_col].mean()
        hard_rim_mean = g_hard[rim_col].mean()
        rim_delta = hard_rim_mean - easy_rim_mean

        rows_maze.append({
            "maze": maze_idx,
            "n_easy": len(g_easy),
            "n_hard": len(g_hard),
            "wasserstein_2d": w2d,
            "js_divergence": jsd,
            "rim_change": rim_delta,
            "prop_easy_rim": easy_rim_mean,
            "prop_hard_rim": hard_rim_mean,
            "mean_easy_path_len": g_easy[trajectory_col].apply(len).mean(),
            "mean_hard_path_len": g_hard[trajectory_col].apply(len).mean(),
        })

    df_maze = pd.DataFrame(rows_maze)

    # ---------------------------------
    # 4) Overall summary
    # ---------------------------------
    g_easy_all = df[df[condition_col] == easy_value]
    g_hard_all = df[df[condition_col] == hard_value]

    # pool all mazes together only if all mazes share same shape
    same_shape = True
    maze_shapes = []
    for maze_idx in sorted(df[maze_col].dropna().unique()):
        maze_shapes.append(maze_to_char_array(test_mazes[int(maze_idx)]).shape)
    if len(set(maze_shapes)) > 1:
        same_shape = False

    overall_dict = {
        "wasserstein_2d_mean_over_subj_maze": df_subj_maze["wasserstein_2d"].mean(),
        "js_divergence_mean_over_subj_maze": df_subj_maze["js_divergence"].mean(),
        "rim_change_mean_over_subj_maze": df_subj_maze["rim_change"].mean(),

        "wasserstein_2d_mean_over_mazes": df_maze["wasserstein_2d"].mean(),
        "js_divergence_mean_over_mazes": df_maze["js_divergence"].mean(),
        "rim_change_mean_over_mazes": df_maze["rim_change"].mean(),

        "prop_easy_rim_overall": g_easy_all[rim_col].mean(),
        "prop_hard_rim_overall": g_hard_all[rim_col].mean(),
        "n_easy_rows": len(g_easy_all),
        "n_hard_rows": len(g_hard_all),
    }

    # optional pooled-overall occupancy metric if all mazes have same shape
    if same_shape and len(g_easy_all) > 0 and len(g_hard_all) > 0:
        maze_shape = maze_to_char_array(test_mazes[int(sorted(df[maze_col].unique())[0])]).shape

        occ_easy_all = occupancy_from_path_list(
            g_easy_all[trajectory_col].tolist(),
            maze_shape,
            swap_xy=swap_xy,
            binary_per_subject=binary_per_path
        )
        occ_hard_all = occupancy_from_path_list(
            g_hard_all[trajectory_col].tolist(),
            maze_shape,
            swap_xy=swap_xy,
            binary_per_subject=binary_per_path
        )

        overall_dict["wasserstein_2d_pooled_allrows"] = wasserstein_2d_exact(
            occ_easy_all, occ_hard_all, metric=wasserstein_metric
        )
        overall_dict["js_divergence_pooled_allrows"] = js_divergence_occupancy(
            occ_easy_all, occ_hard_all
        )
    else:
        overall_dict["wasserstein_2d_pooled_allrows"] = np.nan
        overall_dict["js_divergence_pooled_allrows"] = np.nan

    df_overall = pd.DataFrame([overall_dict])

    return df_subj_maze, df_subject, df_maze, df_overall


df_subj_maze, df_subject, df_maze, df_overall = compute_trace_change_metrics_from_df(
    df_allsubjs_test=df_allsubjs_test,
    test_mazes=test_mazes,
    condition_col="cond",
    easy_value=0,
    hard_value=1,
    subj_col="Subj",
    maze_col="test_maze_orig_idx",
    trajectory_col="trajectory_parsed",
    rim_col="path_on_rim",
    swap_xy=True,
    binary_per_path=False,
    wasserstein_metric="euclidean",
)
```

```python

def build_two_full_rim_routes(max_coord=16):
    """
    Construct the two canonical full rim routes described by you.

    Entry point on rim:  (8, 0)
    Exit point near goal: (8, 14)

    Route A:
      (8,0) -> left along top edge to (0,0)
             -> down left edge to (0,16)
             -> right along bottom edge to (8,16)

    Route B:
      (8,0) -> right along top edge to (16,0)
             -> down right edge to (16,16)
             -> left along bottom edge to (8,16)

    Returns
    -------
    route_left, route_right : sets of (x, y)
    """
    # Left route
    route_left = set()

    # top edge: (8,0) -> (0,0)
    for x in range(0, 9):
        route_left.add((x, 0))

    # left edge: (0,0) -> (0,16)
    for y in range(0, max_coord + 1):
        route_left.add((0, y))

    # bottom edge: (0,16) -> (8,16)
    for x in range(0, 9):
        route_left.add((x, max_coord))

    # Right route
    route_right = set()

    # top edge: (8,0) -> (16,0)
    for x in range(8, max_coord + 1):
        route_right.add((x, 0))

    # right edge: (16,0) -> (16,16)
    for y in range(0, max_coord + 1):
        route_right.add((max_coord, y))

    # bottom edge: (16,16) -> (8,16)
    for x in range(8, max_coord + 1):
        route_right.add((x, max_coord))

    return route_left, route_right

left_rim_path, right_rim_path = build_two_full_rim_routes(max_coord=16)
left_rim_path = left_rim_path|set(start_goal_pos)
right_rim_path = right_rim_path|set(start_goal_pos)
```

```python
import ast
import heapq
import numpy as np
import pandas as pd


# -----------------------------
# Basic parsing helpers
# -----------------------------
def _maybe_literal_eval(x):
    """Safely parse strings like '[8,1]' or "['....', '.##.']"."""
    if isinstance(x, str):
        x = x.strip()
        try:
            return ast.literal_eval(x)
        except Exception:
            return x
    return x


def parse_point(pt):
    """
    Convert a start/goal entry into a tuple (col, row).
    Accepts:
      - [c, r]
      - (c, r)
      - string versions of the above
    """
    pt = _maybe_literal_eval(pt)
    if isinstance(pt, (list, tuple)) and len(pt) == 2:
        return (int(pt[0]), int(pt[1]))
    raise ValueError(f"Could not parse point: {pt}")


def parse_maze_grid(maze):
    """
    Convert maze_grid entry into List[str].
    Accepts:
      - list of strings
      - tuple of strings
      - string representation of list/tuple
    """
    maze = _maybe_literal_eval(maze)

    if isinstance(maze, (list, tuple)):
        maze = [str(row) for row in maze]
        if len(set(len(r) for r in maze)) != 1:
            raise ValueError("Maze rows are not all the same length.")
        return maze

    raise ValueError(f"Could not parse maze_grid: {maze}")


# -----------------------------
# Greedy best-first search
# -----------------------------
def solve_maze_BestFirstSearch(
    maze,
    start=None,
    end=None,
    heuristic="Manhattan",
):
    """
    Perform greedy best-first search on a rectangular maze.

    Parameters
    ----------
    maze : List[str]
        Maze represented as a list of strings of equal length.
        '#' = wall, everything else treated as traversible.
    start : (col, row)
    end : (col, row)
    heuristic : {"Manhattan", "Euclidean"}

    Returns
    -------
    result : dict with keys
        path : List[(col, row)]
        evaluated : List[(col, row)]
        walls : List[(col, row)]
        search_cost : int
        search_found : int
    """
    width = len(maze[0])
    height = len(maze)

    if start is None or end is None:
        raise ValueError("Please provide explicit start and end coordinates.")

    def h(cell):
        c, r = cell
        gc, gr = end
        if heuristic == "Manhattan":
            return abs(c - gc) + abs(r - gr)
        elif heuristic == "Euclidean":
            return np.sqrt((c - gc) ** 2 + (r - gr) ** 2)
        else:
            raise ValueError(f"Unknown heuristic: {heuristic}")

    frontier = [(h(start), start)]
    came_from = {start: None}
    visited = {start}
    evaluated = []
    walls = []

    while frontier:
        _, current = heapq.heappop(frontier)
        evaluated.append(current)

        if current == end:
            break

        c, r = current
        for dc, dr in [(0, -1), (0, 1), (-1, 0), (1, 0)]:
            nc, nr = c + dc, r + dr
            if not (0 <= nc < width and 0 <= nr < height):
                continue

            neighbor = (nc, nr)
            if neighbor in visited:
                continue

            if maze[nr][nc] == "#":
                walls.append(neighbor)
                continue

            visited.add(neighbor)
            came_from[neighbor] = current
            heapq.heappush(frontier, (h(neighbor), neighbor))

    path = []
    if end in came_from:
        node = end
        while node is not None:
            path.append(node)
            node = came_from[node]
        path.reverse()

    search_found = int(len(path) > 0)
    search_cost = len(evaluated) + len(walls)

    return {
        "path": path,
        "evaluated": evaluated,
        "walls": walls,
        "search_cost": search_cost,
        "search_found": search_found,
    }


# -----------------------------
# Classify the BFS path
# -----------------------------
def classify_search_solution_slot(path, left_rim_path, right_rim_path):
    """
    Map a BFS path to:
      0 = left rim solution
      1 = right rim solution
     -1 = neither rim (interior / other)
    """
    if len(path) == 0:
        return 0

    path_set = set(path)

    if path_set.issubset(left_rim_path):
        return 1
    elif path_set.issubset(right_rim_path):
        return 2
    else:
        return 0

def convert_human_solution_slot(x):
    """
    Convert human column path_taken_on_leftorrightrim
    to NumPyro slot convention.
    """
    if x == -1:
        return 0
    elif x == 0:
        return 1
    elif x == 1:
        return 2
    else:
        return 0
    
# -----------------------------
# Main preprocessing function
# -----------------------------
def add_exp1a_precomputations(
    df_human_modeling,
    left_rim_path,
    right_rim_path,
    heuristic="Manhattan",
    add_obs_solution_slot=True,
):
    """
    Add trialwise precomputations for Experiment 1a.

    Parameters
    ----------
    df_human_modeling : pd.DataFrame
        Must contain at least:
            maze_grid
            start_pos
            goal_pos
            left_rim_path_open
            right_rim_path_open
        Optionally:
            path_taken_on_leftorrightrim
    left_rim_path : set of (col, row)
    right_rim_path : set of (col, row)
    heuristic : str
    add_obs_solution_slot : bool
        If True, also add an obs_solution_slot column based on
        path_taken_on_leftorrightrim.

    Returns
    -------
    df_out : pd.DataFrame
        Copy of input dataframe with added columns.
    """
    df = df_human_modeling.copy()

    search_paths = []
    search_costs = []
    search_founds = []
    search_solution_slots = []

    cache_masks = []
    cache_valids = []

    for _, row in df.iterrows():
        maze = parse_maze_grid(row["maze_grid"])
        start = parse_point(row["start_pos"])
        goal = parse_point(row["goal_pos"])

        bfs = solve_maze_BestFirstSearch(
            maze=maze,
            start=start,
            end=goal,
            heuristic=heuristic,
        )

        path = bfs["path"]
        search_cost = bfs["search_cost"]
        search_found = bfs["search_found"]
        search_solution_slot = classify_search_solution_slot(
            path=path,
            left_rim_path=left_rim_path,
            right_rim_path=right_rim_path,
        )

        # In this first-level Exp 1a model, the two reusable solution slots
        # always exist as candidate cache entries.
        cache_mask = [1, 1]

        # Objective validity of the two cached solutions on this trial.
        cache_valid = [
            int(row["left_rim_path_open"]),
            int(row["right_rim_path_open"]),
        ]

        search_paths.append(path)
        search_costs.append(search_cost)
        search_founds.append(search_found)
        search_solution_slots.append(search_solution_slot)

        cache_masks.append(cache_mask)
        cache_valids.append(cache_valid)

    df["search_path"] = search_paths
    df["search_cost"] = search_costs
    df["search_found"] = search_founds
    df["search_solution_slot"] = search_solution_slots

    df["cache_mask"] = cache_masks
    df["cache_valid"] = cache_valids

    # Optional convenience column from the human submission itself:
    #   0 = left rim submitted
    #   1 = right rim submitted
    #  -1 = interior path / search-like non-rim solution
    if add_obs_solution_slot and "path_taken_on_leftorrightrim" in df.columns:
        df["obs_solution_slot"] = df["path_taken_on_leftorrightrim"].apply(
            convert_human_solution_slot
        )

    return df
```

```python
import numpy as np
import pandas as pd


def prepare_numpyro_inputs(
    df_subject_block: pd.DataFrame,
    cond_col: str = "cond",
    cache_mask_col: str = "cache_mask",
    cache_valid_col: str = "cache_valid",
    search_cost_col: str = "search_cost",
    search_found_col: str = "search_found",
    search_solution_slot_col: str = "search_solution_slot",
    obs_solution_slot_col: str = "obs_solution_slot",
    sort_by=("block", "trial"),
    return_metadata: bool = True,
):
    """
    Convert one subject dataframe into NumPy arrays for the NumPyro model.

    Critical:
    ---------
    Trials are sorted in true temporal order, defaulting to:
        block -> trial

    Label convention:
        0 = search
        1 = reuse left
        2 = reuse right
    """
    df = df_subject_block.copy()

    if len(df) == 0:
        raise ValueError("df_subject_block is empty.")

    missing_sort = [c for c in sort_by if c not in df.columns]
    if missing_sort:
        raise ValueError(f"Missing sort columns: {missing_sort}")
    df = df.sort_values(list(sort_by)).reset_index(drop=True)

    required = [
        cond_col,
        cache_mask_col,
        cache_valid_col,
        search_cost_col,
        search_found_col,
        search_solution_slot_col,
        obs_solution_slot_col,
    ]
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    # Condition mapping: smaller value -> 0, larger value -> 1
    cond_vals = pd.unique(df[cond_col])
    cond_vals = [x for x in cond_vals if pd.notnull(x)]
    if len(cond_vals) == 0:
        raise ValueError(f"No non-null values found in '{cond_col}'.")

    cond_vals_sorted = sorted(cond_vals)
    if len(cond_vals_sorted) == 1:
        cond_map = {cond_vals_sorted[0]: 0}
    elif len(cond_vals_sorted) == 2:
        cond_map = {
            cond_vals_sorted[0]: 0,
            cond_vals_sorted[1]: 1,
        }
    else:
        raise ValueError(
            f"Expected 1 or 2 unique condition values in '{cond_col}', got {cond_vals_sorted}."
        )

    cond_idx = df[cond_col].map(cond_map).to_numpy(dtype=np.int32)

    def _to_len2_array(x, name):
        arr = np.asarray(x, dtype=np.float32)
        if arr.shape != (2,):
            raise ValueError(f"{name} entry must have shape (2,), got {arr.shape} with value {x}")
        return arr

    cache_mask = np.stack(
        [_to_len2_array(x, cache_mask_col) for x in df[cache_mask_col].tolist()],
        axis=0
    ).astype(np.float32)

    cache_valid = np.stack(
        [_to_len2_array(x, cache_valid_col) for x in df[cache_valid_col].tolist()],
        axis=0
    ).astype(np.float32)

    search_cost = df[search_cost_col].to_numpy(dtype=np.float32)
    search_found = df[search_found_col].to_numpy(dtype=np.int32)
    search_solution_slot = df[search_solution_slot_col].to_numpy(dtype=np.int32)
    obs_solution_slot = df[obs_solution_slot_col].to_numpy(dtype=np.int32)

    T = len(df)

    if cache_mask.shape != (T, 2):
        raise ValueError(f"cache_mask has shape {cache_mask.shape}, expected ({T}, 2)")
    if cache_valid.shape != (T, 2):
        raise ValueError(f"cache_valid has shape {cache_valid.shape}, expected ({T}, 2)")

    if np.any(~np.isin(search_found, [0, 1])):
        raise ValueError("search_found must contain only 0/1.")
    if np.any(~np.isin(search_solution_slot, [0, 1, 2])):
        raise ValueError("search_solution_slot must contain only {0,1,2}.")
    if np.any(~np.isin(obs_solution_slot, [0, 1, 2])):
        raise ValueError("obs_solution_slot must contain only {0,1,2}.")
    if np.any((cache_mask != 0) & (cache_mask != 1)):
        raise ValueError("cache_mask entries must be 0/1.")
    if np.any((cache_valid != 0) & (cache_valid != 1)):
        raise ValueError("cache_valid entries must be 0/1.")

    out = {
        "cond_idx": cond_idx,
        "cache_mask": cache_mask,
        "cache_valid": cache_valid,
        "search_cost": search_cost,
        "search_found": search_found,
        "search_solution_slot": search_solution_slot,
        "obs_solution_slot": obs_solution_slot,
    }

    if return_metadata:
        meta = {
            "n_trials": T,
            "condition_mapping": cond_map,
            "sort_by": tuple(sort_by),
            "columns_used": {
                "cond": cond_col,
                "cache_mask": cache_mask_col,
                "cache_valid": cache_valid_col,
                "search_cost": search_cost_col,
                "search_found": search_found_col,
                "search_solution_slot": search_solution_slot_col,
                "obs_solution_slot": obs_solution_slot_col,
            },
        }
        for maybe_col in ["Subj", "subj", "subject", "block", "first_cond"]:
            if maybe_col in df.columns:
                vals = df[maybe_col].unique().tolist()
                meta[maybe_col] = vals[0] if len(vals) == 1 else vals

        out["metadata"] = meta

    return out
```

### Numpyro model fits

```python
import jax
import jax.numpy as jnp
import numpyro
import numpyro.distributions as dist


def _safe_normalize(x, eps=1e-12):
    s = jnp.sum(x)
    n = x.shape[0]
    return jnp.where(s > eps, x / s, jnp.ones_like(x) / n)


def _masked_normalize(x, mask, eps=1e-12):
    x = x * mask
    s = jnp.sum(x)
    n_active = jnp.sum(mask)

    uniform = jnp.where(
        n_active > 0,
        mask / jnp.maximum(n_active, 1.0),
        jnp.zeros_like(x),
    )
    return jnp.where(s > eps, x / s, uniform)


def _delta_update_probs(probs, chosen_idx, alpha, mask=None):
    onehot = jax.nn.one_hot(chosen_idx, probs.shape[0], dtype=probs.dtype)
    updated = (1.0 - alpha) * probs + alpha * onehot
    if mask is None:
        return _safe_normalize(updated)
    return _masked_normalize(updated, mask)


def _bernoulli_flip_prob(valid01, lam):
    valid01 = valid01.astype(jnp.float32)
    return valid01 * (1.0 - lam) + (1.0 - valid01) * lam


def colored_maze_reuse_model_marginalized(
    cond_idx,
    cache_mask,
    cache_valid,
    search_cost,
    search_found,
    history_solution_slot=None,   # used only for state updates in observed mode
    obs_solution_slot=None,       # used only for likelihood conditioning
    update_mode="observed",       # "observed" or "sampled"
):
    """
    Label convention:
        0 = final submission from search
        1 = final submission from reuse-left
        2 = final submission from reuse-right

    update_mode:
        "observed" -> state updates use history_solution_slot[t]
        "sampled"  -> state updates use sampled y_t

    Important distinction:
        history_solution_slot : teacher-forcing history
        obs_solution_slot     : optional conditioning target
    """
    if update_mode not in ("observed", "sampled"):
        raise ValueError("update_mode must be 'observed' or 'sampled'")

    T, K = cache_mask.shape
    assert K == 2

    alpha = numpyro.sample("alpha", dist.Beta(2.0, 2.0))
    lam = numpyro.sample("lambda", dist.Beta(2.0, 8.0))

    B_easy = numpyro.sample("B_easy", dist.LogNormal(jnp.log(20.0), 0.75))
    b_frac = numpyro.sample("b_frac", dist.Beta(2.0, 2.0))
    B_hard = numpyro.deterministic("B_hard", B_easy * b_frac)

    lapse = numpyro.sample("lapse", dist.Beta(1.0, 30.0))

    meta_probs = jnp.array([0.5, 0.5], dtype=jnp.float32)  # [search, reuse]
    reuse_probs = _masked_normalize(
        jnp.ones((K,), dtype=jnp.float32),
        cache_mask[0].astype(jnp.float32)
    )

    obs_prob_list = []
    meta_hist = []
    reuse_hist = []

    for t in range(T):
        active = cache_mask[t].astype(jnp.float32)
        valid = cache_valid[t].astype(jnp.float32)

        reuse_probs = _masked_normalize(reuse_probs, active)

        B_t = jnp.where(cond_idx[t] == 0, B_easy, B_hard)
        search_success = ((search_found[t] == 1) & (search_cost[t] <= B_t)).astype(jnp.float32)

        q_valid = _bernoulli_flip_prob(valid, lam) * active
        reuse_success_by_slot = reuse_probs * q_valid
        p_reuse_succeeds = jnp.sum(reuse_success_by_slot)

        p_slot_given_reuse_success = _masked_normalize(reuse_success_by_slot, active)

        p0 = (
            meta_probs[0] * search_success
            + meta_probs[1] * (1.0 - p_reuse_succeeds) * search_success
        )

        p_reuse_total = (
            meta_probs[1]
            + meta_probs[0] * (1.0 - search_success)
        ) * p_reuse_succeeds

        p12 = p_reuse_total * p_slot_given_reuse_success

        p_obs_raw = jnp.concatenate([jnp.array([p0]), p12], axis=0)
        p_obs_core = _safe_normalize(p_obs_raw)
        p_obs = (1.0 - lapse) * p_obs_core + lapse * (jnp.ones(3) / 3.0)
        p_obs = _safe_normalize(p_obs)

        numpyro.deterministic(f"p_obs_{t}", p_obs)
        obs_prob_list.append(p_obs)

        y_obs = None if obs_solution_slot is None else obs_solution_slot[t]
        y_t = numpyro.sample(f"obs_{t}", dist.Categorical(probs=p_obs), obs=y_obs)

        if update_mode == "observed":
            if history_solution_slot is None:
                raise ValueError("update_mode='observed' requires history_solution_slot.")
            y_update = history_solution_slot[t]
        else:
            y_update = y_t

        meta_choice_idx = jnp.where(y_update == 0, 0, 1)
        meta_probs = _delta_update_probs(meta_probs, meta_choice_idx, alpha)

        chosen_cache_idx = jnp.clip(y_update - 1, 0, K - 1)
        reuse_probs_candidate = _delta_update_probs(
            reuse_probs, chosen_cache_idx, alpha, mask=active
        )
        reuse_probs = jnp.where(y_update == 0, reuse_probs, reuse_probs_candidate)

        meta_hist.append(meta_probs)
        reuse_hist.append(reuse_probs)

    numpyro.deterministic("obs_probs", jnp.stack(obs_prob_list))
    numpyro.deterministic("meta_hist", jnp.stack(meta_hist))
    numpyro.deterministic("reuse_hist", jnp.stack(reuse_hist))
```

### Post pred check: condition differences

```python
import numpy as np
import pandas as pd
import jax.numpy as jnp
import jax.random as random
from numpyro.infer import Predictive


def posterior_predictive_compare_by_condition(
    mcmc,
    inputs,
    df_subject,
    cond_col="cond",
    trial_col="trial",
    obs_col="obs_solution_slot",
    n_ppc_draws=None,
    update_mode="observed",
):
    """
    Posterior predictive comparison stratified by condition.

    update_mode:
        "observed" -> sample new labels, but update state using human history
        "sampled"  -> free-running generative simulation
    """
    if update_mode not in ("observed", "sampled"):
        raise ValueError("update_mode must be 'observed' or 'sampled'")

    posterior_samples = mcmc.get_samples()

    if n_ppc_draws is not None:
        n_total = len(next(iter(posterior_samples.values())))
        idx = np.linspace(0, n_total - 1, n_ppc_draws).astype(int)
        posterior_samples = {k: np.asarray(v)[idx] for k, v in posterior_samples.items()}

    T = len(inputs["cond_idx"])

    predictive = Predictive(
        colored_maze_reuse_model_marginalized,
        posterior_samples=posterior_samples,
        return_sites=["obs_probs", "meta_hist", "reuse_hist"] + [f"obs_{t}" for t in range(T)],
    )

    history_for_ppc = jnp.array(inputs["obs_solution_slot"]) if update_mode == "observed" else None

    ppc = predictive(
        random.PRNGKey(123),
        cond_idx=jnp.array(inputs["cond_idx"]),
        cache_mask=jnp.array(inputs["cache_mask"]),
        cache_valid=jnp.array(inputs["cache_valid"]),
        search_cost=jnp.array(inputs["search_cost"]),
        search_found=jnp.array(inputs["search_found"]),
        history_solution_slot=history_for_ppc,
        obs_solution_slot=None,     # critical: do NOT condition during PPC
        update_mode=update_mode,
    )

    sampled_labels = np.stack([np.asarray(ppc[f"obs_{t}"]) for t in range(T)], axis=1)
    obs_probs_post = np.asarray(ppc["obs_probs"])
    meta_hist_post = np.asarray(ppc["meta_hist"])
    reuse_hist_post = np.asarray(ppc["reuse_hist"])

    obs_probs_mean = obs_probs_post.mean(axis=0)

    sort_by = inputs.get("metadata", {}).get("sort_by", ("block", "trial"))
    df_aug = df_subject.sort_values(list(sort_by)).reset_index(drop=True).copy()

    if len(df_aug) != T:
        raise ValueError(f"Dataframe has {len(df_aug)} rows, but model inputs have {T} trials.")

    df_aug["pp_mean_p_search"] = obs_probs_mean[:, 0]
    df_aug["pp_mean_p_reuse_left"] = obs_probs_mean[:, 1]
    df_aug["pp_mean_p_reuse_right"] = obs_probs_mean[:, 2]
    df_aug["pp_mean_p_reuse_total"] = obs_probs_mean[:, 1] + obs_probs_mean[:, 2]

    # Actual human summaries
    actual_rows = []
    for cond_val, d in df_aug.groupby(cond_col, sort=True):
        y = d[obs_col].to_numpy()
        n = len(y)
        n_search = np.sum(y == 0)
        n_reuse = np.sum(np.isin(y, [1, 2]))
        n_left = np.sum(y == 1)
        n_right = np.sum(y == 2)

        actual_rows.append({
            cond_col: cond_val,
            "n_trials": n,
            "p_search": n_search / n,
            "p_reuse": n_reuse / n,
            "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
            "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
            "n_search": n_search,
            "n_reuse": n_reuse,
            "n_left": n_left,
            "n_right": n_right,
        })

    actual_by_cond = pd.DataFrame(actual_rows).sort_values(cond_col).reset_index(drop=True)

    # PPC summaries by draw and condition
    cond_values = df_aug[cond_col].to_numpy()
    unique_conds = np.sort(df_aug[cond_col].unique())

    ppc_rows = []
    for s in range(sampled_labels.shape[0]):
        ys = sampled_labels[s]
        for cond_val in unique_conds:
            mask = (cond_values == cond_val)
            yc = ys[mask]
            n = len(yc)

            n_search = np.sum(yc == 0)
            n_reuse = np.sum(np.isin(yc, [1, 2]))
            n_left = np.sum(yc == 1)
            n_right = np.sum(yc == 2)

            ppc_rows.append({
                "draw": s,
                cond_col: cond_val,
                "n_trials": n,
                "p_search": n_search / n,
                "p_reuse": n_reuse / n,
                "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
                "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
                "n_search": n_search,
                "n_reuse": n_reuse,
                "n_left": n_left,
                "n_right": n_right,
            })

    ppc_by_cond_draw = pd.DataFrame(ppc_rows)

    metrics = ["p_search", "p_reuse", "p_left_given_reuse", "p_right_given_reuse"]
    comp_rows = []

    for cond_val in unique_conds:
        actual_row = actual_by_cond.loc[actual_by_cond[cond_col] == cond_val].iloc[0]
        ppc_sub = ppc_by_cond_draw.loc[ppc_by_cond_draw[cond_col] == cond_val]

        for metric in metrics:
            vals = ppc_sub[metric].dropna().to_numpy()
            comp_rows.append({
                cond_col: cond_val,
                "metric": metric,
                "actual": actual_row[metric],
                "ppc_mean": np.mean(vals),
                "ppc_sd": np.std(vals),
                "ppc_q05": np.quantile(vals, 0.05),
                "ppc_q50": np.quantile(vals, 0.50),
                "ppc_q95": np.quantile(vals, 0.95),
                "update_mode": update_mode,
            })

    comparison_by_cond = pd.DataFrame(comp_rows).sort_values([cond_col, "metric"]).reset_index(drop=True)

    return {
        "sampled_labels": sampled_labels,
        "obs_probs_post": obs_probs_post,
        "meta_hist_post": meta_hist_post,
        "reuse_hist_post": reuse_hist_post,
        "df_aug": df_aug,
        "actual_by_cond": actual_by_cond,
        "ppc_by_cond_draw": ppc_by_cond_draw,
        "comparison_by_cond": comparison_by_cond,
        "update_mode": update_mode,
    }
```

```python
import matplotlib.pyplot as plt


def plot_ppc_by_condition(ppc_out, cond_col="cond"):
    actual_by_cond = ppc_out["actual_by_cond"]
    ppc_by_cond_draw = ppc_out["ppc_by_cond_draw"]
    update_mode = ppc_out.get("update_mode", "unknown")

    metrics = [
        ("p_search", "P(search)"),
        ("p_reuse", "P(reuse)"),
        ("p_left_given_reuse", "P(left | reuse)"),
        ("p_right_given_reuse", "P(right | reuse)"),
    ]

    cond_vals = np.sort(actual_by_cond[cond_col].unique())
    nrows = len(cond_vals)
    ncols = len(metrics)

    fig, axes = plt.subplots(nrows, ncols, figsize=(4 * ncols, 3.2 * nrows), squeeze=False)

    for i, cond_val in enumerate(cond_vals):
        act_row = actual_by_cond.loc[actual_by_cond[cond_col] == cond_val].iloc[0]
        ppc_sub = ppc_by_cond_draw.loc[ppc_by_cond_draw[cond_col] == cond_val]

        for j, (metric, title) in enumerate(metrics):
            ax = axes[i, j]
            vals = ppc_sub[metric].dropna().to_numpy()
            ax.hist(vals, bins=30)
            ax.axvline(act_row[metric], color="r", linestyle="--")
            ax.set_title(f"{title}\ncond={cond_val}, mode={update_mode}")
            ax.set_xlabel("Posterior predictive")
            ax.set_ylabel("Cond "+str(cond_val)+"\nCount")

    plt.tight_layout()
    plt.show()
    
    
import matplotlib.pyplot as plt


def plot_ppc_by_condition(ppc_out, cond_col="cond"):
    """
    Histogram PPC distributions by condition, with actual human value overlaid.
    """
    actual_by_cond = ppc_out["actual_by_cond"]
    ppc_by_cond_draw = ppc_out["ppc_by_cond_draw"]

    metrics = [
        ("p_search", "P(search)"),
        ("p_reuse", "P(reuse)"),
        ("p_left_given_reuse", "P(left | reuse)"),
        ("p_right_given_reuse", "P(right | reuse)"),
    ]

    cond_vals = np.sort(actual_by_cond[cond_col].unique())
    nrows = len(cond_vals)
    ncols = len(metrics)

    fig, axes = plt.subplots(nrows, ncols, figsize=(4 * ncols, 3.2 * nrows), squeeze=False)

    for i, cond_val in enumerate(cond_vals):
        act_row = actual_by_cond.loc[actual_by_cond[cond_col] == cond_val].iloc[0]
        ppc_sub = ppc_by_cond_draw.loc[ppc_by_cond_draw[cond_col] == cond_val]

        for j, (metric, title) in enumerate(metrics):
            ax = axes[i, j]
            vals = ppc_sub[metric].dropna().to_numpy()
            ax.hist(vals, bins=30)
            ax.axvline(act_row[metric], color="r", linestyle="--")
            ax.set_title(f"{title}\ncond={cond_val}")
            ax.set_xlabel("Posterior predictive")
            ax.set_ylabel("Count")

    plt.tight_layout()
    plt.show()
    
    
def plot_ppc_condition_differences(diff_out):
    """
    Plot PPC distribution of condition differences with actual difference overlaid.
    """
    comparison_diff = diff_out["comparison_diff"]
    ppc_diff_draw = diff_out["ppc_diff_draw"]

    metrics = [
        ("p_search", "Δ P(search)"),
        ("p_reuse", "Δ P(reuse)"),
        ("p_left_given_reuse", "Δ P(left | reuse)"),
        ("p_right_given_reuse", "Δ P(right | reuse)"),
    ]

    fig, axes = plt.subplots(1, len(metrics), figsize=(4 * len(metrics), 3.5), squeeze=False)
    axes = axes[0]

    for ax, (metric, title) in zip(axes, metrics):
        vals = ppc_diff_draw.loc[ppc_diff_draw["metric"] == metric, "ppc_diff"].dropna().to_numpy()
        actual_val = comparison_diff.loc[comparison_diff["metric"] == metric, "actual_diff"].iloc[0]

        ax.hist(vals, bins=30)
        ax.axvline(actual_val, color="r", linestyle="--")
        ax.set_title(title)
        ax.set_xlabel("Posterior predictive difference")
        ax.set_ylabel("Count")

    plt.tight_layout()
    plt.show()
    
def plot_trialwise_posterior_probs_by_condition(ppc_out, cond_col="cond", trial_col="trial"):
    """
    Plot trialwise posterior mean label probabilities separately by condition.
    """
    df_aug = ppc_out["df_aug"].copy()
    cond_vals = np.sort(df_aug[cond_col].unique())

    fig, axes = plt.subplots(len(cond_vals), 1, figsize=(10, 3.5 * len(cond_vals)), squeeze=False)
    axes = axes[:, 0]

    for ax, cond_val in zip(axes, cond_vals):
        d = df_aug.loc[df_aug[cond_col] == cond_val].sort_values(trial_col)

        ax.plot(d[trial_col], d["pp_mean_p_search"], label="search")
        ax.plot(d[trial_col], d["pp_mean_p_reuse_left"], label="reuse_left")
        ax.plot(d[trial_col], d["pp_mean_p_reuse_right"], label="reuse_right")
        ax.plot(d[trial_col], d["pp_mean_p_reuse_total"], label="reuse_total", linestyle="--")

        ax.set_title(f"Posterior mean trialwise probabilities, cond={cond_val}")
        ax.set_xlabel("Trial")
        ax.set_ylabel("Cond "+str(cond_val)+"\nProbability")
        ax.legend()

    plt.tight_layout()
    plt.show()
    
import matplotlib.pyplot as plt
def summarize_condition_differences(ppc_out, cond_col="cond"):
    """
    Compute actual and posterior predictive condition differences.
    Difference is defined as:
        larger condition value - smaller condition value
    """
    actual_by_cond = ppc_out["actual_by_cond"].copy()
    ppc_by_cond_draw = ppc_out["ppc_by_cond_draw"].copy()

    cond_vals = np.sort(actual_by_cond[cond_col].unique())
    if len(cond_vals) != 2:
        raise ValueError(f"Expected exactly 2 conditions, got {cond_vals}")

    c0, c1 = cond_vals[0], cond_vals[1]
    metrics = ["p_search", "p_reuse", "p_left_given_reuse", "p_right_given_reuse"]

    # Actual differences
    a0 = actual_by_cond.loc[actual_by_cond[cond_col] == c0].iloc[0]
    a1 = actual_by_cond.loc[actual_by_cond[cond_col] == c1].iloc[0]

    actual_diff_rows = []
    for metric in metrics:
        actual_diff_rows.append({
            "metric": metric,
            "actual_diff": a1[metric] - a0[metric],
            "cond_low": c0,
            "cond_high": c1,
        })
    actual_diff_df = pd.DataFrame(actual_diff_rows)

    # PPC differences by draw
    diff_rows = []
    for draw, dsub in ppc_by_cond_draw.groupby("draw"):
        r0 = dsub.loc[dsub[cond_col] == c0].iloc[0]
        r1 = dsub.loc[dsub[cond_col] == c1].iloc[0]

        for metric in metrics:
            diff_rows.append({
                "draw": draw,
                "metric": metric,
                "ppc_diff": r1[metric] - r0[metric],
                "cond_low": c0,
                "cond_high": c1,
            })

    ppc_diff_draw = pd.DataFrame(diff_rows)

    # Summaries
    comp_rows = []
    for metric in metrics:
        vals = ppc_diff_draw.loc[ppc_diff_draw["metric"] == metric, "ppc_diff"].dropna().to_numpy()
        actual_val = actual_diff_df.loc[actual_diff_df["metric"] == metric, "actual_diff"].iloc[0]

        comp_rows.append({
            "metric": metric,
            "actual_diff": actual_val,
            "ppc_mean_diff": np.mean(vals),
            "ppc_sd_diff": np.std(vals),
            "ppc_q05_diff": np.quantile(vals, 0.05),
            "ppc_q50_diff": np.quantile(vals, 0.50),
            "ppc_q95_diff": np.quantile(vals, 0.95),
        })

    comparison_diff = pd.DataFrame(comp_rows)
    return {
        "actual_diff_df": actual_diff_df,
        "ppc_diff_draw": ppc_diff_draw,
        "comparison_diff": comparison_diff,
    }

def plot_ppc_by_condition(ppc_out, cond_col="cond"):
    """
    Histogram PPC distributions by condition, with actual human value overlaid.
    """
    actual_by_cond = ppc_out["actual_by_cond"]
    ppc_by_cond_draw = ppc_out["ppc_by_cond_draw"]

    metrics = [
        ("p_search", "P(search)"),
        ("p_reuse", "P(reuse)"),
        ("p_left_given_reuse", "P(left | reuse)"),
        ("p_right_given_reuse", "P(right | reuse)"),
    ]

    cond_vals = np.sort(actual_by_cond[cond_col].unique())
    nrows = len(cond_vals)
    ncols = len(metrics)

    fig, axes = plt.subplots(nrows, ncols, figsize=(4 * ncols, 3.2 * nrows), squeeze=False)

    for i, cond_val in enumerate(cond_vals):
        act_row = actual_by_cond.loc[actual_by_cond[cond_col] == cond_val].iloc[0]
        ppc_sub = ppc_by_cond_draw.loc[ppc_by_cond_draw[cond_col] == cond_val]

        for j, (metric, title) in enumerate(metrics):
            ax = axes[i, j]
            vals = ppc_sub[metric].dropna().to_numpy()
            ax.hist(vals, bins=30)
            ax.axvline(act_row[metric], color="r", linestyle="--")
            ax.set_title(f"{title}\ncond={cond_val}")
            ax.set_xlabel("Posterior predictive")
            ax.set_ylabel("Count")

    plt.tight_layout()
    plt.show()
    
    
def plot_ppc_condition_differences(diff_out):
    """
    Plot PPC distribution of condition differences with actual difference overlaid.
    """
    comparison_diff = diff_out["comparison_diff"]
    ppc_diff_draw = diff_out["ppc_diff_draw"]

    metrics = [
        ("p_search", "Δ P(search)"),
        ("p_reuse", "Δ P(reuse)"),
        ("p_left_given_reuse", "Δ P(left | reuse)"),
        ("p_right_given_reuse", "Δ P(right | reuse)"),
    ]

    fig, axes = plt.subplots(1, len(metrics), figsize=(4 * len(metrics), 3.5), squeeze=False)
    axes = axes[0]

    for ax, (metric, title) in zip(axes, metrics):
        vals = ppc_diff_draw.loc[ppc_diff_draw["metric"] == metric, "ppc_diff"].dropna().to_numpy()
        actual_val = comparison_diff.loc[comparison_diff["metric"] == metric, "actual_diff"].iloc[0]

        ax.hist(vals, bins=30)
        ax.axvline(actual_val, color="r", linestyle="--")
        ax.set_title(title)
        ax.set_xlabel("Posterior predictive difference")
        ax.set_ylabel("Count")

    plt.tight_layout()
    plt.show()
    
def plot_trialwise_posterior_probs_by_condition(ppc_out, cond_col="cond", trial_col="trial"):
    """
    Plot trialwise posterior mean label probabilities separately by condition.
    """
    df_aug = ppc_out["df_aug"].copy()
    cond_vals = np.sort(df_aug[cond_col].unique())

    fig, axes = plt.subplots(len(cond_vals), 1, figsize=(10, 3.5 * len(cond_vals)), squeeze=False)
    axes = axes[:, 0]

    for ax, cond_val in zip(axes, cond_vals):
        d = df_aug.loc[df_aug[cond_col] == cond_val].sort_values(trial_col)

        ax.plot(d[trial_col], d["pp_mean_p_search"], label="search")
        ax.plot(d[trial_col], d["pp_mean_p_reuse_left"], label="reuse_left")
        ax.plot(d[trial_col], d["pp_mean_p_reuse_right"], label="reuse_right")
        ax.plot(d[trial_col], d["pp_mean_p_reuse_total"], label="reuse_total", linestyle="--")

        ax.set_title(f"Posterior mean trialwise probabilities, cond={cond_val}")
        ax.set_xlabel("Trial")
        ax.set_ylabel("Probability")
        ax.legend()

    plt.tight_layout()
    plt.show()
```

```python
ppc_obs = posterior_predictive_compare_by_condition(
    mcmc=mcmc,
    inputs=inputs,
    df_subject=df_one,
    cond_col="cond",
    trial_col="trial",
    obs_col="obs_solution_slot",
    n_ppc_draws=1000,
    update_mode="sampled", # "observed", "sampled"
)

diff_obs = summarize_condition_differences(ppc_obs, cond_col="cond")

print(ppc_obs["comparison_by_cond"])
print(diff_obs["comparison_diff"])

plot_ppc_by_condition(ppc_obs, cond_col="cond")
plot_ppc_condition_differences(diff_obs)
```

### Post pred check: both conditions together

```python
# Posterior predictive samples---do they match the human data being fitted on?

import numpy as np
import pandas as pd
import jax.numpy as jnp
import jax.random as random
from numpyro.infer import Predictive


def posterior_predictive_compare_frequencies(
    mcmc,
    inputs,
    df_one,
    n_ppc_draws=None,
    block_col = "block",
    trial_col="trial",
    obs_col="obs_solution_slot",
):
    """
    Generate posterior predictive samples from the fitted marginalized model
    and compare:
      1) meta choice frequency: search vs reuse
      2) conditional reuse choice frequency: left vs right given reuse

    Label convention:
      0 = search
      1 = reuse-left
      2 = reuse-right

    Parameters
    ----------
    mcmc : numpyro.infer.MCMC
        Fitted MCMC object.
    inputs : dict
        Output of prepare_numpyro_inputs(df_one).
    df_one : pd.DataFrame
        The original dataframe for this subject-block.
    n_ppc_draws : int or None
        If None, uses all posterior draws. Otherwise subsamples that many.
    trial_col : str
        Trial column in df_one.
    obs_col : str
        Observed label column in df_one.

    Returns
    -------
    out : dict
        Contains:
          - ppc_samples_labels : [S, T] simulated labels
          - actual_summary : dict
          - ppc_summary_df : dataframe across posterior predictive draws
          - comparison_df : tidy dataframe with actual vs posterior predictive summary
          - df_aug : df_one sorted with posterior mean trialwise label probs
    """
    posterior_samples = mcmc.get_samples()

    # Optionally subsample posterior draws for PPC
    if n_ppc_draws is not None:
        n_total = len(next(iter(posterior_samples.values())))
        idx = np.linspace(0, n_total - 1, n_ppc_draws).astype(int)
        posterior_samples = {k: np.asarray(v)[idx] for k, v in posterior_samples.items()}

    # Posterior predictive:
    # pass obs_solution_slot=None so the model samples new labels
    predictive = Predictive(
        colored_maze_reuse_model_marginalized,
        posterior_samples=posterior_samples,
        return_sites=["obs_probs"] + [f"obs_{t}" for t in range(len(inputs["cond_idx"]))],
    )

    ppc = predictive(
        random.PRNGKey(123),
        cond_idx=jnp.array(inputs["cond_idx"]),
        cache_mask=jnp.array(inputs["cache_mask"]),
        cache_valid=jnp.array(inputs["cache_valid"]),
        search_cost=jnp.array(inputs["search_cost"]),
        search_found=jnp.array(inputs["search_found"]),
        obs_solution_slot=None,
    )

    # Collect sampled labels into [S, T]
    T = len(inputs["cond_idx"])
    sampled_labels = np.stack([np.asarray(ppc[f"obs_{t}"]) for t in range(T)], axis=1)

    # Posterior mean trialwise probabilities, useful for inspection
    obs_probs_post = np.asarray(ppc["obs_probs"])  # [S, T, 3]
    obs_probs_mean = obs_probs_post.mean(axis=0)   # [T, 3]

    # -------------------------
    # Actual human frequencies
    # -------------------------
    df_sorted = df_one.sort_values((block_col, trial_col)).reset_index(drop=True).copy()
    y = df_sorted[obs_col].to_numpy()

    actual_n_search = np.sum(y == 0)
    actual_n_reuse = np.sum(np.isin(y, [1, 2]))
    actual_n_left = np.sum(y == 1)
    actual_n_right = np.sum(y == 2)
    actual_T = len(y)

    actual_summary = {
        "n_trials": actual_T,
        "p_search": actual_n_search / actual_T,
        "p_reuse": actual_n_reuse / actual_T,
        "p_left_given_reuse": (
            actual_n_left / actual_n_reuse if actual_n_reuse > 0 else np.nan
        ),
        "p_right_given_reuse": (
            actual_n_right / actual_n_reuse if actual_n_reuse > 0 else np.nan
        ),
        "n_search": actual_n_search,
        "n_reuse": actual_n_reuse,
        "n_left": actual_n_left,
        "n_right": actual_n_right,
    }

    # -------------------------
    # PPC frequencies by draw
    # -------------------------
    rows = []
    for s in range(sampled_labels.shape[0]):
        ys = sampled_labels[s]

        n_search = np.sum(ys == 0)
        n_reuse = np.sum(np.isin(ys, [1, 2]))
        n_left = np.sum(ys == 1)
        n_right = np.sum(ys == 2)

        rows.append({
            "draw": s,
            "p_search": n_search / T,
            "p_reuse": n_reuse / T,
            "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
            "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
            "n_search": n_search,
            "n_reuse": n_reuse,
            "n_left": n_left,
            "n_right": n_right,
        })

    ppc_summary_df = pd.DataFrame(rows)

    # Summarize PPC distribution
    metrics = ["p_search", "p_reuse", "p_left_given_reuse", "p_right_given_reuse"]
    comp_rows = []
    for metric in metrics:
        vals = ppc_summary_df[metric].dropna().to_numpy()
        comp_rows.append({
            "metric": metric,
            "actual": actual_summary[metric],
            "ppc_mean": np.mean(vals),
            "ppc_sd": np.std(vals),
            "ppc_q05": np.quantile(vals, 0.05),
            "ppc_q50": np.quantile(vals, 0.50),
            "ppc_q95": np.quantile(vals, 0.95),
        })

    comparison_df = pd.DataFrame(comp_rows)

    # Augment df with posterior mean trialwise probs
    df_aug = df_sorted.copy()
    df_aug["pp_mean_p_search"] = obs_probs_mean[:, 0]
    df_aug["pp_mean_p_reuse_left"] = obs_probs_mean[:, 1]
    df_aug["pp_mean_p_reuse_right"] = obs_probs_mean[:, 2]
    df_aug["pp_mean_p_reuse_total"] = obs_probs_mean[:, 1] + obs_probs_mean[:, 2]

    return {
        "ppc_samples_labels": sampled_labels,   # [S, T]
        "actual_summary": actual_summary,
        "ppc_summary_df": ppc_summary_df,
        "comparison_df": comparison_df,
        "df_aug": df_aug,
    }
```

```python
ppc_out = posterior_predictive_compare_frequencies(
    mcmc=mcmc,
    inputs=inputs,
    df_one=df_one,
    n_ppc_draws=1000,   # optional
)

print("Actual summary:")
print(ppc_out["actual_summary"])
print()

print("Comparison:")
print(ppc_out["comparison_df"])
```

```python
comparison_df = ppc_out["comparison_df"].copy()
comparison_df["actual"] = comparison_df["actual"].round(3)
comparison_df["ppc_mean"] = comparison_df["ppc_mean"].round(3)
comparison_df["ppc_sd"] = comparison_df["ppc_sd"].round(3)
comparison_df["ppc_q05"] = comparison_df["ppc_q05"].round(3)
comparison_df["ppc_q50"] = comparison_df["ppc_q50"].round(3)
comparison_df["ppc_q95"] = comparison_df["ppc_q95"].round(3)
print(comparison_df)
```

#### Post pred check: Compare by cond

```python
import numpy as np
import pandas as pd
import jax.numpy as jnp
import jax.random as random
from numpyro.infer import Predictive


def posterior_predictive_compare_by_condition(
    mcmc,
    inputs,
    df_subject,
    cond_col="cond",
    trial_col="trial",
    obs_col="obs_solution_slot",
    n_ppc_draws=None,
):
    """
    Posterior predictive comparison stratified by condition.

    Assumes model was fit on one subject across both conditions together.

    Label convention
    ----------------
    0 = search
    1 = reuse-left
    2 = reuse-right

    Returns
    -------
    out : dict
        sampled_labels : [S, T]
        df_aug : trial-level dataframe with posterior mean probs
        actual_by_cond : dataframe
        ppc_by_cond_draw : dataframe, one row per posterior draw x condition
        comparison_by_cond : dataframe
    """
    posterior_samples = mcmc.get_samples()

    if n_ppc_draws is not None:
        n_total = len(next(iter(posterior_samples.values())))
        idx = np.linspace(0, n_total - 1, n_ppc_draws).astype(int)
        posterior_samples = {k: np.asarray(v)[idx] for k, v in posterior_samples.items()}

    T = len(inputs["cond_idx"])

    predictive = Predictive(
        colored_maze_reuse_model_marginalized,
        posterior_samples=posterior_samples,
        return_sites=["obs_probs"] + [f"obs_{t}" for t in range(T)],
    )

    ppc = predictive(
        random.PRNGKey(123),
        cond_idx=jnp.array(inputs["cond_idx"]),
        cache_mask=jnp.array(inputs["cache_mask"]),
        cache_valid=jnp.array(inputs["cache_valid"]),
        search_cost=jnp.array(inputs["search_cost"]),
        search_found=jnp.array(inputs["search_found"]),
        obs_solution_slot=None,
    )

    sampled_labels = np.stack([np.asarray(ppc[f"obs_{t}"]) for t in range(T)], axis=1)  # [S, T]
    obs_probs_post = np.asarray(ppc["obs_probs"])  # [S, T, 3]
    obs_probs_mean = obs_probs_post.mean(axis=0)   # [T, 3]

    # Sort dataframe to match model order
    sort_by = inputs.get("metadata", {}).get("sort_by", ("block", "trial"))
    df_aug = df_subject.sort_values(list(sort_by)).reset_index(drop=True).copy()

    # Sanity check on trial count
    if len(df_aug) != T:
        raise ValueError(f"Dataframe has {len(df_aug)} rows, but model inputs have {T} trials.")

    # Attach posterior mean probabilities
    df_aug["pp_mean_p_search"] = obs_probs_mean[:, 0]
    df_aug["pp_mean_p_reuse_left"] = obs_probs_mean[:, 1]
    df_aug["pp_mean_p_reuse_right"] = obs_probs_mean[:, 2]
    df_aug["pp_mean_p_reuse_total"] = obs_probs_mean[:, 1] + obs_probs_mean[:, 2]

    # -------------------------
    # Actual human summaries by condition
    # -------------------------
    actual_rows = []
    for cond_val, d in df_aug.groupby(cond_col, sort=True):
        y = d[obs_col].to_numpy()
        n = len(y)
        n_search = np.sum(y == 0)
        n_reuse = np.sum(np.isin(y, [1, 2]))
        n_left = np.sum(y == 1)
        n_right = np.sum(y == 2)

        actual_rows.append({
            cond_col: cond_val,
            "n_trials": n,
            "p_search": n_search / n,
            "p_reuse": n_reuse / n,
            "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
            "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
            "n_search": n_search,
            "n_reuse": n_reuse,
            "n_left": n_left,
            "n_right": n_right,
        })

    actual_by_cond = pd.DataFrame(actual_rows).sort_values(cond_col).reset_index(drop=True)

    # -------------------------
    # PPC summaries by condition and draw
    # -------------------------
    cond_values = df_aug[cond_col].to_numpy()
    unique_conds = np.sort(df_aug[cond_col].unique())

    ppc_rows = []
    for s in range(sampled_labels.shape[0]):
        ys = sampled_labels[s]
        for cond_val in unique_conds:
            mask = (cond_values == cond_val)
            yc = ys[mask]
            n = len(yc)

            n_search = np.sum(yc == 0)
            n_reuse = np.sum(np.isin(yc, [1, 2]))
            n_left = np.sum(yc == 1)
            n_right = np.sum(yc == 2)

            ppc_rows.append({
                "draw": s,
                cond_col: cond_val,
                "n_trials": n,
                "p_search": n_search / n,
                "p_reuse": n_reuse / n,
                "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
                "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
                "n_search": n_search,
                "n_reuse": n_reuse,
                "n_left": n_left,
                "n_right": n_right,
            })

    ppc_by_cond_draw = pd.DataFrame(ppc_rows)

    # -------------------------
    # Compare actual vs PPC by condition
    # -------------------------
    metrics = ["p_search", "p_reuse", "p_left_given_reuse", "p_right_given_reuse"]
    comp_rows = []

    for cond_val in unique_conds:
        actual_row = actual_by_cond.loc[actual_by_cond[cond_col] == cond_val].iloc[0]
        ppc_sub = ppc_by_cond_draw.loc[ppc_by_cond_draw[cond_col] == cond_val]

        for metric in metrics:
            vals = ppc_sub[metric].dropna().to_numpy()
            comp_rows.append({
                cond_col: cond_val,
                "metric": metric,
                "actual": actual_row[metric],
                "ppc_mean": np.mean(vals),
                "ppc_sd": np.std(vals),
                "ppc_q05": np.quantile(vals, 0.05),
                "ppc_q50": np.quantile(vals, 0.50),
                "ppc_q95": np.quantile(vals, 0.95),
            })

    comparison_by_cond = pd.DataFrame(comp_rows).sort_values([cond_col, "metric"]).reset_index(drop=True)

    return {
        "sampled_labels": sampled_labels,
        "df_aug": df_aug,
        "actual_by_cond": actual_by_cond,
        "ppc_by_cond_draw": ppc_by_cond_draw,
        "comparison_by_cond": comparison_by_cond,
    }
```

```python
def summarize_condition_differences(ppc_out, cond_col="cond"):
    """
    Compute actual and posterior predictive condition differences.
    Difference is defined as:
        larger condition value - smaller condition value
    """
    actual_by_cond = ppc_out["actual_by_cond"].copy()
    ppc_by_cond_draw = ppc_out["ppc_by_cond_draw"].copy()

    cond_vals = np.sort(actual_by_cond[cond_col].unique())
    if len(cond_vals) != 2:
        raise ValueError(f"Expected exactly 2 conditions, got {cond_vals}")

    c0, c1 = cond_vals[0], cond_vals[1]
    metrics = ["p_search", "p_reuse", "p_left_given_reuse", "p_right_given_reuse"]

    # Actual differences
    a0 = actual_by_cond.loc[actual_by_cond[cond_col] == c0].iloc[0]
    a1 = actual_by_cond.loc[actual_by_cond[cond_col] == c1].iloc[0]

    actual_diff_rows = []
    for metric in metrics:
        actual_diff_rows.append({
            "metric": metric,
            "actual_diff": a1[metric] - a0[metric],
            "cond_low": c0,
            "cond_high": c1,
        })
    actual_diff_df = pd.DataFrame(actual_diff_rows)

    # PPC differences by draw
    diff_rows = []
    for draw, dsub in ppc_by_cond_draw.groupby("draw"):
        r0 = dsub.loc[dsub[cond_col] == c0].iloc[0]
        r1 = dsub.loc[dsub[cond_col] == c1].iloc[0]

        for metric in metrics:
            diff_rows.append({
                "draw": draw,
                "metric": metric,
                "ppc_diff": r1[metric] - r0[metric],
                "cond_low": c0,
                "cond_high": c1,
            })

    ppc_diff_draw = pd.DataFrame(diff_rows)

    # Summaries
    comp_rows = []
    for metric in metrics:
        vals = ppc_diff_draw.loc[ppc_diff_draw["metric"] == metric, "ppc_diff"].dropna().to_numpy()
        actual_val = actual_diff_df.loc[actual_diff_df["metric"] == metric, "actual_diff"].iloc[0]

        comp_rows.append({
            "metric": metric,
            "actual_diff": actual_val,
            "ppc_mean_diff": np.mean(vals),
            "ppc_sd_diff": np.std(vals),
            "ppc_q05_diff": np.quantile(vals, 0.05),
            "ppc_q50_diff": np.quantile(vals, 0.50),
            "ppc_q95_diff": np.quantile(vals, 0.95),
        })

    comparison_diff = pd.DataFrame(comp_rows)
    return {
        "actual_diff_df": actual_diff_df,
        "ppc_diff_draw": ppc_diff_draw,
        "comparison_diff": comparison_diff,
    }
```

```python
import matplotlib.pyplot as plt


def plot_ppc_by_condition(ppc_out, cond_col="cond"):
    """
    Histogram PPC distributions by condition, with actual human value overlaid.
    """
    actual_by_cond = ppc_out["actual_by_cond"]
    ppc_by_cond_draw = ppc_out["ppc_by_cond_draw"]

    metrics = [
        ("p_search", "P(search)"),
        ("p_reuse", "P(reuse)"),
        ("p_left_given_reuse", "P(left | reuse)"),
        ("p_right_given_reuse", "P(right | reuse)"),
    ]

    cond_vals = np.sort(actual_by_cond[cond_col].unique())
    nrows = len(cond_vals)
    ncols = len(metrics)

    fig, axes = plt.subplots(nrows, ncols, figsize=(4 * ncols, 3.2 * nrows), squeeze=False)

    for i, cond_val in enumerate(cond_vals):
        act_row = actual_by_cond.loc[actual_by_cond[cond_col] == cond_val].iloc[0]
        ppc_sub = ppc_by_cond_draw.loc[ppc_by_cond_draw[cond_col] == cond_val]

        for j, (metric, title) in enumerate(metrics):
            ax = axes[i, j]
            vals = ppc_sub[metric].dropna().to_numpy()
            ax.hist(vals, bins=30)
            ax.axvline(act_row[metric], color="r", linestyle="--")
            ax.set_title(f"{title}\ncond={cond_val}")
            ax.set_xlabel("Posterior predictive")
            ax.set_ylabel("Count")

    plt.tight_layout()
    plt.show()
    
    
def plot_ppc_condition_differences(diff_out):
    """
    Plot PPC distribution of condition differences with actual difference overlaid.
    """
    comparison_diff = diff_out["comparison_diff"]
    ppc_diff_draw = diff_out["ppc_diff_draw"]

    metrics = [
        ("p_search", "Δ P(search)"),
        ("p_reuse", "Δ P(reuse)"),
        ("p_left_given_reuse", "Δ P(left | reuse)"),
        ("p_right_given_reuse", "Δ P(right | reuse)"),
    ]

    fig, axes = plt.subplots(1, len(metrics), figsize=(4 * len(metrics), 3.5), squeeze=False)
    axes = axes[0]

    for ax, (metric, title) in zip(axes, metrics):
        vals = ppc_diff_draw.loc[ppc_diff_draw["metric"] == metric, "ppc_diff"].dropna().to_numpy()
        actual_val = comparison_diff.loc[comparison_diff["metric"] == metric, "actual_diff"].iloc[0]

        ax.hist(vals, bins=30)
        ax.axvline(actual_val, color="r", linestyle="--")
        ax.set_title(title)
        ax.set_xlabel("Posterior predictive difference")
        ax.set_ylabel("Count")

    plt.tight_layout()
    plt.show()
    
def plot_trialwise_posterior_probs_by_condition(ppc_out, cond_col="cond", trial_col="trial"):
    """
    Plot trialwise posterior mean label probabilities separately by condition.
    """
    df_aug = ppc_out["df_aug"].copy()
    cond_vals = np.sort(df_aug[cond_col].unique())

    fig, axes = plt.subplots(len(cond_vals), 1, figsize=(10, 3.5 * len(cond_vals)), squeeze=False)
    axes = axes[:, 0]

    for ax, cond_val in zip(axes, cond_vals):
        d = df_aug.loc[df_aug[cond_col] == cond_val].sort_values(trial_col)

        ax.plot(d[trial_col], d["pp_mean_p_search"], label="search")
        ax.plot(d[trial_col], d["pp_mean_p_reuse_left"], label="reuse_left")
        ax.plot(d[trial_col], d["pp_mean_p_reuse_right"], label="reuse_right")
        ax.plot(d[trial_col], d["pp_mean_p_reuse_total"], label="reuse_total", linestyle="--")

        ax.set_title(f"Posterior mean trialwise probabilities, cond={cond_val}")
        ax.set_xlabel("Trial")
        ax.set_ylabel("Probability")
        ax.legend()

    plt.tight_layout()
    plt.show()
```
