## Numpyro model

```python

def build_two_full_rim_routes(max_coord=16):
    """
    Construct the two canonical full rim routes described by you.

    Entry point on rim:  (8, 0)
    Exit point near goal: (8, 14)

    Route A:
      (8,0) -> left along top edge to (0,0)
             -> down left edge to (0,16)
             -> right along bottom edge to (8,16)

    Route B:
      (8,0) -> right along top edge to (16,0)
             -> down right edge to (16,16)
             -> left along bottom edge to (8,16)

    Returns
    -------
    route_left, route_right : sets of (x, y)
    """
    # Left route
    route_left = set()

    # top edge: (8,0) -> (0,0)
    for x in range(0, 9):
        route_left.add((x, 0))

    # left edge: (0,0) -> (0,16)
    for y in range(0, max_coord + 1):
        route_left.add((0, y))

    # bottom edge: (0,16) -> (8,16)
    for x in range(0, 9):
        route_left.add((x, max_coord))

    # Right route
    route_right = set()

    # top edge: (8,0) -> (16,0)
    for x in range(8, max_coord + 1):
        route_right.add((x, 0))

    # right edge: (16,0) -> (16,16)
    for y in range(0, max_coord + 1):
        route_right.add((max_coord, y))

    # bottom edge: (16,16) -> (8,16)
    for x in range(8, max_coord + 1):
        route_right.add((x, max_coord))

    return route_left, route_right

left_rim_path, right_rim_path = build_two_full_rim_routes(max_coord=16)
left_rim_path = left_rim_path|set(start_goal_pos)
right_rim_path = right_rim_path|set(start_goal_pos)
rim_path_length = len(left_rim_path)
```

```python

def prepare_numpyro_inputs(
    df_subject_block: pd.DataFrame,
    subj_col: str = "subj_idx",
    cond_col: str = "cond",
    cache_mask_col: str = "cache_mask",
    cache_valid_col: str = "cache_valid",
    search_path_len_col: str = "search_path_len",
    search_cost_col: str = "search_cost",
    search_found_col: str = "search_found",
    search_solution_slot_col: str = "search_solution_slot",
    obs_solution_slot_col: str = "obs_solution_slot",
    sort_by=("subj_idx","block", "trial"),
    return_metadata: bool = True,
):
    """
    Convert one subject dataframe into NumPy arrays for the NumPyro model.

    Critical:
    ---------
    Trials are sorted in true temporal order, defaulting to:
        block -> trial

    Label convention:
        0 = search
        1 = reuse left
        2 = reuse right
    """
    df = df_subject_block.copy()

    if len(df) == 0:
        raise ValueError("df_subject_block is empty.")

    missing_sort = [c for c in sort_by if c not in df.columns]
    if missing_sort:
        raise ValueError(f"Missing sort columns: {missing_sort}")
    df = df.sort_values(list(sort_by)).reset_index(drop=True)

    required = [
        subj_col,
        cond_col,
        cache_mask_col,
        cache_valid_col,
        search_path_len_col,
        search_cost_col,
        search_found_col,
        search_solution_slot_col,
        obs_solution_slot_col,
    ]
    missing = [c for c in required if c not in df.columns]
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    # Condition mapping: smaller value -> 0, larger value -> 1
    cond_vals = pd.unique(df[cond_col])
    cond_vals = [x for x in cond_vals if pd.notnull(x)]
    if len(cond_vals) == 0:
        raise ValueError(f"No non-null values found in '{cond_col}'.")

    cond_vals_sorted = sorted(cond_vals)
    if len(cond_vals_sorted) == 1:
        cond_map = {cond_vals_sorted[0]: 0}
    elif len(cond_vals_sorted) == 2:
        cond_map = {
            cond_vals_sorted[0]: 0,
            cond_vals_sorted[1]: 1,
        }
    else:
        raise ValueError(
            f"Expected 1 or 2 unique condition values in '{cond_col}', got {cond_vals_sorted}."
        )

    cond_idx = df[cond_col].map(cond_map).to_numpy(dtype=np.int32)

    def _to_len2_array(x, name):
        arr = np.asarray(x, dtype=np.float32)
        if arr.shape != (2,):
            raise ValueError(f"{name} entry must have shape (2,), got {arr.shape} with value {x}")
        return arr

    cache_mask = np.stack(
        [_to_len2_array(x, cache_mask_col) for x in df[cache_mask_col].tolist()],
        axis=0
    ).astype(np.float32)

    cache_valid = np.stack(
        [_to_len2_array(x, cache_valid_col) for x in df[cache_valid_col].tolist()],
        axis=0
    ).astype(np.float32)

    subj = df[subj_col].to_numpy(dtype=np.int32)
    search_path_len = df[search_path_len_col].to_numpy(dtype=np.float32)
    search_cost = df[search_cost_col].to_numpy(dtype=np.float32)
    search_found = df[search_found_col].to_numpy(dtype=np.int32)
    search_solution_slot = df[search_solution_slot_col].to_numpy(dtype=np.int32)
    obs_solution_slot = df[obs_solution_slot_col].to_numpy(dtype=np.int32)

    T = len(df)

    if cache_mask.shape != (T, 2):
        raise ValueError(f"cache_mask has shape {cache_mask.shape}, expected ({T}, 2)")
    if cache_valid.shape != (T, 2):
        raise ValueError(f"cache_valid has shape {cache_valid.shape}, expected ({T}, 2)")

    if np.any(~np.isin(search_found, [0, 1])):
        raise ValueError("search_found must contain only 0/1.")
    if np.any(~np.isin(search_solution_slot, [0, 1, 2])):
        raise ValueError("search_solution_slot must contain only {0,1,2}.")
    if np.any(~np.isin(obs_solution_slot, [0, 1, 2])):
        raise ValueError("obs_solution_slot must contain only {0,1,2}.")
    if np.any((cache_mask != 0) & (cache_mask != 1)):
        raise ValueError("cache_mask entries must be 0/1.")
    if np.any((cache_valid != 0) & (cache_valid != 1)):
        raise ValueError("cache_valid entries must be 0/1.")

    out = {
        "subj_idx": subj,
        "cond_idx": cond_idx,
        "cache_mask": cache_mask,
        "cache_valid": cache_valid,
        "search_path_len": search_path_len,
        "search_cost": search_cost,
        "search_found": search_found,
        "search_solution_slot": search_solution_slot,
        "obs_solution_slot": obs_solution_slot,
    }

    if return_metadata:
        meta = {
            "n_trials": T,
            "condition_mapping": cond_map,
            "sort_by": tuple(sort_by),
            "columns_used": {
                "subj": subj_col,
                "cond": cond_col,
                "cache_mask": cache_mask_col,
                "cache_valid": cache_valid_col,
                "search_path_len": search_path_len,
                "search_cost": search_cost_col,
                "search_found": search_found_col,
                "search_solution_slot": search_solution_slot_col,
                "obs_solution_slot": obs_solution_slot_col,
            },
        }
        for maybe_col in ["Subj", "subj", "subject", "block", "first_cond"]:
            if maybe_col in df.columns:
                vals = df[maybe_col].unique().tolist()
                meta[maybe_col] = vals[0] if len(vals) == 1 else vals

        out["metadata"] = meta

    return out
```

### Numpyro model fits

```python
import jax
import jax.numpy as jnp
import numpyro
import numpyro.distributions as dist


def colored_maze_static_utility_hierarchical(
    subj_idx,              # shape [T], integer-coded 0..S-1
    cond_idx,              # shape [T], integer-coded 0..C-1
    cache_valid,           # shape [T, 2], 1 if rim path valid/open, 0 otherwise
    rim_path_length,       # shape [T, 2], length of left/right rim path
    EU_search,             # shape [T], expected utility of search
    EN_search,             # shape [T], expected search cost
    obs_solution_slot=None,# shape [T], 0=search, 1=left, 2=right
    n_subj=None,
    n_cond=None,
    L_budget = 70,
):
    """
    Static hierarchical softmax model.

    Action coding:
        0 = search
        1 = reuse-left
        2 = reuse-right

    Utilities:
        U_left  = valid_left  * (N_budget - rim_len_left)
        U_right = valid_right * (N_budget - rim_len_right)
        U_search = EU_search - lambda_{s,c} * EN_search

    Choice:
        y_t ~ Categorical(softmax(beta_s * [U_search, U_left, U_right]))

    Hierarchical parameters:
        beta_s > 0                       subject-specific inverse temperature
        lambda_{s,c} > 0                subject x condition search-cost weight
    """

    if n_subj is None:
        raise ValueError("n_subj must be provided as a Python int.")
    if n_cond is None:
        raise ValueError("n_cond must be provided as a Python int.")

    T = subj_idx.shape[0]
    if cache_valid.shape != rim_path_length.shape:
        raise ValueError("cache_valid and rim_path_length must have same shape [T, 2].")
    if cache_valid.shape[1] != 2:
        raise ValueError("Expected exactly 2 rim paths: left and right.")

    # -----------------------------
    # Hierarchical beta_s > 0
    # log(beta_s) ~ Normal(mu_beta, sigma_beta)
    # -----------------------------
    mu_log_beta = numpyro.sample("mu_log_beta", dist.Normal(0.0, 1.0))
    sigma_log_beta = numpyro.sample("sigma_log_beta", dist.HalfNormal(0.7))

    with numpyro.plate("subj_plate_beta", n_subj):
        log_beta_subj = numpyro.sample(
            "log_beta_subj",
            dist.Normal(mu_log_beta, sigma_log_beta)
        )

    beta_subj = numpyro.deterministic("beta_subj", jnp.exp(log_beta_subj))

    # -----------------------------
    # Hierarchical lambda_{s,c} > 0
    #
    # Non-centered decomposition:
    #   group mean by condition
    #   subject deviation shared through z
    # -----------------------------
    with numpyro.plate("cond_plate_lambda", n_cond):
        mu_log_lambda_cond = numpyro.sample(
            "mu_log_lambda_cond",
            dist.Normal(0.0, 1.0)
        )
        sigma_log_lambda_cond = numpyro.sample(
            "sigma_log_lambda_cond",
            dist.HalfNormal(0.7)
        )

    with numpyro.plate("subj_plate_lambda", n_subj):
        with numpyro.plate("cond_plate_lambda_subj", n_cond):
            z_lambda = numpyro.sample("z_lambda", dist.Normal(0.0, 1.0))

    z_lambda = jnp.swapaxes(z_lambda, 0, 1)   # now shape should be (n_subj, n_cond)

    log_lambda_subj_cond = (
        mu_log_lambda_cond[None, :] +
        sigma_log_lambda_cond[None, :] * z_lambda
    )

    lambda_subj_cond = numpyro.deterministic(
        "lambda_subj_cond",
        jnp.exp(log_lambda_subj_cond)
    )

    # -----------------------------
    # Trial-wise utilities
    # -----------------------------
    valid = cache_valid.astype(jnp.float32)

    U_left = valid[:, 0] * (L_budget - rim_path_length[:, 0])
    U_right = valid[:, 1] * (L_budget - rim_path_length[:, 1])

    beta_t = beta_subj[subj_idx]
    lambda_t = lambda_subj_cond[subj_idx, cond_idx]

    U_search = EU_search - lambda_t * EN_search

    utilities = jnp.stack([U_search, U_left, U_right], axis=1)
    numpyro.deterministic("utilities", utilities)

    logits = beta_t[:, None] * utilities
    numpyro.deterministic("logits", logits)

    probs = jax.nn.softmax(logits, axis=1)
    numpyro.deterministic("obs_probs", probs)

    with numpyro.plate("trials", T):
        numpyro.sample(
            "obs",
            dist.Categorical(logits=logits),
            obs=obs_solution_slot
        )
        
        
import jax
import jax.numpy as jnp
import numpyro
import numpyro.distributions as dist


def colored_maze_utility_hierarchical(
    subj_idx,              # shape [T], integer-coded 0..S-1
    cond_idx,              # shape [T], integer-coded 0..C-1
    cache_valid,           # shape [T, 2], 1 if rim path valid/open, 0 otherwise
    rim_path_length,       # shape [T, 2], length of left/right rim path
    EU_search,             # shape [T], expected utility of search
    EN_search,             # shape [T], expected search cost
    obs_solution_slot=None,# shape [T], 0=search, 1=left, 2=right
    n_subj=None,
    n_cond=None,
    L_budget = 70,
):
    """
    Static hierarchical softmax model.

    Action coding:
        0 = search
        1 = reuse-left
        2 = reuse-right

    Utilities:
        U_left  = valid_left  * (N_budget - rim_len_left)
        U_right = valid_right * (N_budget - rim_len_right)
        U_search = EU_search - lambda_{s,c} * EN_search

    Choice:
        y_t ~ Categorical(softmax(beta_s * [U_search, U_left, U_right]))

    Hierarchical parameters:
        beta_s > 0                       subject-specific inverse temperature
        lambda_{s,c} > 0                subject x condition search-cost weight
    """

    if n_subj is None:
        raise ValueError("n_subj must be provided as a Python int.")
    if n_cond is None:
        raise ValueError("n_cond must be provided as a Python int.")

    T = subj_idx.shape[0]
    if cache_valid.shape != rim_path_length.shape:
        raise ValueError("cache_valid and rim_path_length must have same shape [T, 2].")
    if cache_valid.shape[1] != 2:
        raise ValueError("Expected exactly 2 rim paths: left and right.")

    # -----------------------------
    # Hierarchical beta_s > 0
    # log(beta_s) ~ Normal(mu_beta, sigma_beta)
    # -----------------------------
    mu_log_beta = numpyro.sample("mu_log_beta", dist.Normal(0.0, 1.0))
    sigma_log_beta = numpyro.sample("sigma_log_beta", dist.HalfNormal(0.7))

    with numpyro.plate("subj_plate_beta", n_subj):
        log_beta_subj = numpyro.sample(
            "log_beta_subj",
            dist.Normal(mu_log_beta, sigma_log_beta)
        )

    beta_subj = numpyro.deterministic("beta_subj", jnp.exp(log_beta_subj))

    # -----------------------------
    # Hierarchical lambda_{s,c} > 0
    #
    # Non-centered decomposition:
    #   group mean by condition
    #   subject deviation shared through z
    # -----------------------------
    with numpyro.plate("cond_plate_lambda", n_cond):
        mu_log_lambda_cond = numpyro.sample(
            "mu_log_lambda_cond",
            dist.Normal(0.0, 1.0)
        )
        sigma_log_lambda_cond = numpyro.sample(
            "sigma_log_lambda_cond",
            dist.HalfNormal(0.7)
        )

    with numpyro.plate("subj_plate_lambda", n_subj):
        with numpyro.plate("cond_plate_lambda_subj", n_cond):
            z_lambda = numpyro.sample("z_lambda", dist.Normal(0.0, 1.0))

    z_lambda = jnp.swapaxes(z_lambda, 0, 1)   # now shape should be (n_subj, n_cond)

    log_lambda_subj_cond = (
        mu_log_lambda_cond[None, :] +
        sigma_log_lambda_cond[None, :] * z_lambda
    )

    lambda_subj_cond = numpyro.deterministic(
        "lambda_subj_cond",
        jnp.exp(log_lambda_subj_cond)
    )

    # -----------------------------
    # Trial-wise utilities
    # -----------------------------
    valid = cache_valid.astype(jnp.float32)

#     U_left = valid[:, 0] * (L_budget - rim_path_length[:, 0])
#     U_right = valid[:, 1] * (L_budget - rim_path_length[:, 1])
    
    # How much subjects dislike a blocked rim path.
    with numpyro.plate("subj_plate_kappa", n_subj):
        kappa_subj = numpyro.sample("kappa_subj", dist.Gamma(8,1/3))  # mean 5

    kappa_t = kappa_subj[subj_idx]

    U_left = jnp.where(
        cache_valid[:, 0] == 1,
        L_budget - rim_path_length[:, 0],
        -kappa_t
    )

    U_right = jnp.where(
        cache_valid[:, 1] == 1,
        L_budget - rim_path_length[:, 1],
        -kappa_t
    )

    beta_t = beta_subj[subj_idx]
    lambda_t = lambda_subj_cond[subj_idx, cond_idx]
    
    
    # Specific to search: subjective estimates on the expected utility of search, and the #squares needed to be inspected.
    with numpyro.plate("subj_plate_Lsearch", n_subj):
        L_search_subj = numpyro.sample(
            "L_search_subj",
            dist.Gamma(8,1/3)
        )
    with numpyro.plate("subj_plate_Nsearch", n_subj):
        N_search_subj = numpyro.sample(
            "N_search_subj",
            dist.Gamma(30,1/4)
        )

    L_search_t = L_search_subj[subj_idx]      # shape [T]
    N_search_t = N_search_subj[subj_idx]      # shape [T]

    U_search = (L_budget - L_search_t) - lambda_t * N_search_t   # shape [T]

    utilities = jnp.stack([U_search, U_left, U_right], axis=1)
    numpyro.deterministic("utilities", utilities)

    logits = beta_t[:, None] * utilities
    numpyro.deterministic("logits", logits)

    probs = jax.nn.softmax(logits, axis=1)
    numpyro.deterministic("obs_probs", probs)

    with numpyro.plate("trials", T):
        numpyro.sample(
            "obs",
            dist.Categorical(logits=logits),
            obs=obs_solution_slot
        )
```

```python
import jax
import jax.numpy as jnp
import numpyro
import numpyro.distributions as dist


def colored_maze_utility_hierarchical_v2(
    subj_idx,              # shape [T], integer-coded 0..S-1
    cond_idx,              # shape [T], integer-coded 0..C-1
    cache_valid,           # shape [T, 2], 1 if rim path valid/open, 0 otherwise
    rim_path_length,       # shape [T, 2], length of left/right rim path
    EU_search,             # kept for API compatibility; not used in this version
    EN_search,             # kept for API compatibility; not used in this version
    obs_solution_slot=None,# shape [T], 0=search, 1=left, 2=right
    n_subj=None,
    n_cond=None,
    L_budget=70.0,
):
    """
    Hierarchical static softmax model with:
      - subject-level inverse temperature beta_s
      - subject-level blocked-path aversion kappa_s
      - subject-level subjective search burden N_search_s
      - subject-level subjective expected interior path length L_search_s
      - subject-level baseline reuse_left bias rho_s_left
      - subject-level baseline reuse_right bias rho_s_right
      - subject baseline + hard-condition shift in log lambda

    Action coding:
        0 = search
        1 = reuse-left
        2 = reuse-right

    Notes:
        EU_search and EN_search are accepted only to preserve the same
        function signature as earlier models. They are not used here.
    """
    if n_subj is None:
        raise ValueError("n_subj must be provided as a Python int.")
    if n_cond is None:
        raise ValueError("n_cond must be provided as a Python int.")
    if n_cond != 2:
        raise ValueError("This version assumes exactly 2 conditions.")

    T = subj_idx.shape[0]
    if cache_valid.shape != rim_path_length.shape:
        raise ValueError("cache_valid and rim_path_length must have same shape [T, 2].")
    if cache_valid.shape[1] != 2:
        raise ValueError("Expected exactly 2 rim paths: left and right.")

    # -----------------------------
    # Subject-level inverse temperature
    # -----------------------------
    mu_log_beta = numpyro.sample("mu_log_beta", dist.Normal(-1.0, 0.5))
    sigma_log_beta = numpyro.sample("sigma_log_beta", dist.HalfNormal(0.4))
    z_beta = numpyro.sample("z_beta", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_beta_subj = mu_log_beta + sigma_log_beta * z_beta
    beta_subj = numpyro.deterministic("beta_subj", jnp.exp(log_beta_subj))

    # -----------------------------
    # Subject-level blocked-path aversion
    # -----------------------------
    mu_log_kappa = numpyro.sample("mu_log_kappa", dist.Normal(2.5, 0.5))
    sigma_log_kappa = numpyro.sample("sigma_log_kappa", dist.HalfNormal(0.4))
    z_kappa = numpyro.sample("z_kappa", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_kappa_subj = mu_log_kappa + sigma_log_kappa * z_kappa
    kappa_subj = numpyro.deterministic("kappa_subj", jnp.exp(log_kappa_subj))

    # -----------------------------
    # Subject-level baseline reuse bias
    # Positive rho means cached reuse is more attractive overall
    # -----------------------------
    mu_rho_left = numpyro.sample("mu_rho_left", dist.Normal(0.0, 10.0))
    sigma_rho_left = numpyro.sample("sigma_rho_left", dist.HalfNormal(5.0))
    z_rho_left = numpyro.sample("z_rho_left", dist.Normal(0.0, 1.0).expand([n_subj]))
    rho_left_subj = numpyro.deterministic("rho_subj_left", mu_rho_left + sigma_rho_left * z_rho_left)
    
    mu_rho_right = numpyro.sample("mu_rho_right", dist.Normal(0.0, 10.0))
    sigma_rho_right = numpyro.sample("sigma_rho_right", dist.HalfNormal(5.0))
    z_rho_right = numpyro.sample("z_rho_right", dist.Normal(0.0, 1.0).expand([n_subj]))
    rho_right_subj = numpyro.deterministic("rho_subj_right", mu_rho_right + sigma_rho_right * z_rho_right)


#     # -----------------------------
#     # Subjective search burden: N_search_s
#     # -----------------------------
#     mu_log_Nsearch = numpyro.sample("mu_log_Nsearch", dist.Normal(jnp.log(20.0), 0.5))
#     sigma_log_Nsearch = numpyro.sample("sigma_log_Nsearch", dist.HalfNormal(0.4))
#     z_Nsearch = numpyro.sample("z_Nsearch", dist.Normal(0.0, 1.0).expand([n_subj]))
#     log_Nsearch_subj = mu_log_Nsearch + sigma_log_Nsearch * z_Nsearch
#     N_search_subj = numpyro.deterministic("N_search_subj", jnp.exp(log_Nsearch_subj))

#     # -----------------------------
#     # Shared subjective expected interior path length
#     # -----------------------------
# #     L_search = numpyro.sample("L_search", dist.Gamma(12.0, 0.5))  # mean 24
#     mu_log_Lsearch = numpyro.sample("mu_log_Lsearch", dist.Normal(jnp.log(20.0), 0.4))
#     sigma_log_Lsearch = numpyro.sample("sigma_log_Lsearch", dist.HalfNormal(0.3))
#     z_Lsearch = numpyro.sample("z_Lsearch", dist.Normal(0, 1).expand([n_subj]))
#     log_Lsearch_subj = mu_log_Lsearch + sigma_log_Lsearch * z_Lsearch
#     L_search_subj = numpyro.deterministic("L_search_subj", jnp.exp(log_Nsearch_subj))


    # -----------------------------
    # Subject baseline + hard-condition shift in log lambda
    # This keeps your original idea that condition affects search cost
    # through lambda_c * N_search
    # -----------------------------
    mu_log_lambda = numpyro.sample("mu_log_lambda", dist.Normal(0.0, 0.5))
    sigma_log_lambda = numpyro.sample("sigma_log_lambda", dist.HalfNormal(0.3))
    z_lambda_base = numpyro.sample("z_lambda_base", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_lambda_base_subj = mu_log_lambda + sigma_log_lambda * z_lambda_base

    mu_delta_lambda = numpyro.sample("mu_delta_lambda", dist.Normal(0.0, 0.4))
    sigma_delta_lambda = numpyro.sample("sigma_delta_lambda", dist.HalfNormal(0.25))
    z_delta_lambda = numpyro.sample("z_delta_lambda", dist.Normal(0.0, 1.0).expand([n_subj]))
    delta_lambda_subj = mu_delta_lambda + sigma_delta_lambda * z_delta_lambda

    # -----------------------------
    # Trial-level indexing
    # -----------------------------
    beta_t = beta_subj[subj_idx]
    kappa_t = kappa_subj[subj_idx]
    rho_left_t = rho_left_subj[subj_idx]
    rho_right_t = rho_right_subj[subj_idx]
#     N_search_t = N_search_subj[subj_idx]
#     L_search_t = L_search_subj[subj_idx]

    log_lambda_t = log_lambda_base_subj[subj_idx] + delta_lambda_subj[subj_idx] * (cond_idx == 1)
    lambda_t = numpyro.deterministic("lambda_t", jnp.exp(log_lambda_t))

    # -----------------------------
    # Reuse utilities
    # Base value if open; blocked penalty if closed
    # Then add subject reuse bias and hard-condition reuse shift
    # -----------------------------
    U_left_base = jnp.where(
        cache_valid[:, 0] == 1,
        L_budget - rim_path_length[:, 0],
        -kappa_t
    )

    U_right_base = jnp.where(
        cache_valid[:, 1] == 1,
        L_budget - rim_path_length[:, 1],
        -kappa_t
    )

    U_left = U_left_base + rho_left_t
    U_right = U_right_base + rho_right_t

    # -----------------------------
    # Search utility
    # Keep simple: expected path gain minus condition-weighted inspection burden
    # -----------------------------
    U_search = EU_search - lambda_t * EN_search

    utilities = jnp.stack([U_search, U_left, U_right], axis=1)
    logits = beta_t[:, None] * utilities

    numpyro.deterministic("utilities", utilities)
    numpyro.deterministic("logits", logits)
    numpyro.deterministic("obs_probs", jax.nn.softmax(logits, axis=1))

    with numpyro.plate("trials", T):
        numpyro.sample("obs", dist.Categorical(logits=logits), obs=obs_solution_slot)
```

```python
def colored_maze_utility_hierarchical_v3(
    subj_idx,
    cond_idx,
    cache_valid,
    rim_path_length,
    EU_search,             # shape [T], expected utility of search
    EN_search,             # shape [T], expected search cost
    obs_solution_slot=None,
    n_subj=None,
    n_cond=None,
    L_budget=70.0,
):
    if n_subj is None:
        raise ValueError("n_subj must be provided as a Python int.")
    if n_cond != 2:
        raise ValueError("This parameterization assumes exactly 2 conditions.")

    T = subj_idx.shape[0]

    # beta_s
    mu_log_beta = numpyro.sample("mu_log_beta", dist.Normal(-1.0, 0.5))
    sigma_log_beta = numpyro.sample("sigma_log_beta", dist.HalfNormal(0.4))
    z_beta = numpyro.sample("z_beta", dist.Normal(0, 1).expand([n_subj]))
    log_beta_subj = mu_log_beta + sigma_log_beta * z_beta
    beta_subj = numpyro.deterministic("beta_subj", jnp.exp(log_beta_subj))

    # kappa_s (blocked-path aversion)
    mu_log_kappa = numpyro.sample("mu_log_kappa", dist.Normal(2.5, 0.4))
    sigma_log_kappa = numpyro.sample("sigma_log_kappa", dist.HalfNormal(0.3))
    z_kappa = numpyro.sample("z_kappa", dist.Normal(0, 1).expand([n_subj]))
    log_kappa_subj = mu_log_kappa + sigma_log_kappa * z_kappa
    kappa_subj = numpyro.deterministic("kappa_subj", jnp.exp(log_kappa_subj))

    # subjective search burden: fit ONLY N_search_s for now
    mu_log_Nsearch = numpyro.sample("mu_log_Nsearch", dist.Normal(jnp.log(20.0), 0.4))
    sigma_log_Nsearch = numpyro.sample("sigma_log_Nsearch", dist.HalfNormal(0.3))
    z_Nsearch = numpyro.sample("z_Nsearch", dist.Normal(0, 1).expand([n_subj]))
    log_Nsearch_subj = mu_log_Nsearch + sigma_log_Nsearch * z_Nsearch
    N_search_subj = numpyro.deterministic("N_search_subj", jnp.exp(log_Nsearch_subj))

    # subjective expected interior path length: FIXED for now
#     L_search = numpyro.sample("L_search", dist.Gamma(12.0, 0.5))  # mean 24
    mu_log_Lsearch = numpyro.sample("mu_log_Lsearch", dist.Normal(jnp.log(20.0), 0.4))
    sigma_log_Lsearch = numpyro.sample("sigma_log_Lsearch", dist.HalfNormal(0.3))
    z_Lsearch = numpyro.sample("z_Lsearch", dist.Normal(0, 1).expand([n_subj]))
    log_Lsearch_subj = mu_log_Lsearch + sigma_log_Lsearch * z_Lsearch
    L_search_subj = numpyro.deterministic("L_search_subj", jnp.exp(log_Lsearch_subj))

    # lambda_s baseline + condition effect
    mu_log_lambda = numpyro.sample("mu_log_lambda", dist.Normal(0.0, 0.5))
    sigma_log_lambda = numpyro.sample("sigma_log_lambda", dist.HalfNormal(0.3))
    z_lambda_base = numpyro.sample("z_lambda_base", dist.Normal(0, 1).expand([n_subj]))
    log_lambda_base_subj = mu_log_lambda + sigma_log_lambda * z_lambda_base

    # population hard-condition effect + subject deviations
    mu_delta_lambda = numpyro.sample("mu_delta_lambda", dist.Normal(0.0, 0.4))
    sigma_delta_lambda = numpyro.sample("sigma_delta_lambda", dist.HalfNormal(0.2))
    z_delta_lambda = numpyro.sample("z_delta_lambda", dist.Normal(0, 1).expand([n_subj]))
    delta_lambda_subj = mu_delta_lambda + sigma_delta_lambda * z_delta_lambda

    log_lambda_t = log_lambda_base_subj[subj_idx] + delta_lambda_subj[subj_idx] * (cond_idx == 1)
    lambda_t = numpyro.deterministic("lambda_t", jnp.exp(log_lambda_t))

    beta_t = beta_subj[subj_idx]
    kappa_t = kappa_subj[subj_idx]
    L_search_t = L_search_subj[subj_idx]
    N_search_t = N_search_subj[subj_idx]

    U_left = jnp.where(
        cache_valid[:, 0] == 1,
        L_budget - rim_path_length[:, 0],
        -kappa_t
    )

    U_right = jnp.where(
        cache_valid[:, 1] == 1,
        L_budget - rim_path_length[:, 1],
        -kappa_t
    )

    U_search = (L_budget - L_search_t) - lambda_t * N_search_t

    utilities = jnp.stack([U_search, U_left, U_right], axis=1)
    logits = beta_t[:, None] * utilities

    numpyro.deterministic("utilities", utilities)
    numpyro.deterministic("obs_probs", jax.nn.softmax(logits, axis=1))

    with numpyro.plate("trials", T):
        numpyro.sample("obs", dist.Categorical(logits=logits), obs=obs_solution_slot)
```

```python
import jax
import jax.numpy as jnp
import numpyro
import numpyro.distributions as dist


def colored_maze_utility_hierarchical_v4(
    subj_idx,              # shape [T], integer-coded 0..S-1
    cond_idx,              # shape [T], integer-coded 0..C-1
    cache_valid,           # shape [T, 2], 1 if rim path valid/open, 0 otherwise
    rim_path_length,       # shape [T, 2], length of left/right rim path
    EU_search,             # kept for API compatibility; not used in this version
    EN_search,             # kept for API compatibility; not used in this version
    obs_solution_slot=None,# shape [T], 0=search, 1=left, 2=right
    n_subj=None,
    n_cond=None,
    L_budget=70.0,
):
    """
    Hierarchical static softmax model with:
      - subject-level inverse temperature beta_s
      - subject-level blocked-path aversion kappa_s
      - subject-level subjective search burden N_search_s
      - shared subjective expected interior path length L_search
      - subject-level baseline reuse bias rho_s
      - subject-level hard-condition reuse bonus delta_reuse_s
      - subject baseline + hard-condition shift in log lambda

    Action coding:
        0 = search
        1 = reuse-left
        2 = reuse-right

    Notes:
        EU_search and EN_search are accepted only to preserve the same
        function signature as earlier models. They are not used here.
    """
    if n_subj is None:
        raise ValueError("n_subj must be provided as a Python int.")
    if n_cond is None:
        raise ValueError("n_cond must be provided as a Python int.")
    if n_cond != 2:
        raise ValueError("This version assumes exactly 2 conditions.")

    T = subj_idx.shape[0]
    if cache_valid.shape != rim_path_length.shape:
        raise ValueError("cache_valid and rim_path_length must have same shape [T, 2].")
    if cache_valid.shape[1] != 2:
        raise ValueError("Expected exactly 2 rim paths: left and right.")

    # -----------------------------
    # Subject-level inverse temperature
    # -----------------------------
    mu_log_beta = numpyro.sample("mu_log_beta", dist.Normal(-1.0, 0.5))
    sigma_log_beta = numpyro.sample("sigma_log_beta", dist.HalfNormal(0.4))
    z_beta = numpyro.sample("z_beta", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_beta_subj = mu_log_beta + sigma_log_beta * z_beta
    beta_subj = numpyro.deterministic("beta_subj", jnp.exp(log_beta_subj))

    # -----------------------------
    # Subject-level blocked-path aversion
    # -----------------------------
    mu_log_kappa = numpyro.sample("mu_log_kappa", dist.Normal(2.5, 0.5))
    sigma_log_kappa = numpyro.sample("sigma_log_kappa", dist.HalfNormal(0.4))
    z_kappa = numpyro.sample("z_kappa", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_kappa_subj = mu_log_kappa + sigma_log_kappa * z_kappa
    kappa_subj = numpyro.deterministic("kappa_subj", jnp.exp(log_kappa_subj))

    # -----------------------------
    # Subject-level baseline reuse bias
    # Positive rho means cached reuse is more attractive overall
    # -----------------------------
    mu_rho = numpyro.sample("mu_rho", dist.Normal(0.0, 10.0))
    sigma_rho = numpyro.sample("sigma_rho", dist.HalfNormal(5.0))
    z_rho = numpyro.sample("z_rho", dist.Normal(0.0, 1.0).expand([n_subj]))
    rho_subj = numpyro.deterministic("rho_subj", mu_rho + sigma_rho * z_rho)

    # -----------------------------
    # Subject-level hard-condition reuse bonus
    # Only applies when cond_idx == 1
    # Positive delta means more reuse in hard condition
    # -----------------------------
    mu_delta_reuse = numpyro.sample("mu_delta_reuse", dist.Normal(0.0, 10.0))
    sigma_delta_reuse = numpyro.sample("sigma_delta_reuse", dist.HalfNormal(5.0))
    z_delta_reuse = numpyro.sample("z_delta_reuse", dist.Normal(0.0, 1.0).expand([n_subj]))
    delta_reuse_subj = numpyro.deterministic(
        "delta_reuse_subj",
        mu_delta_reuse + sigma_delta_reuse * z_delta_reuse
    )

    # -----------------------------
    # Subjective search burden: N_search_s
    # -----------------------------
    mu_log_Nsearch = numpyro.sample("mu_log_Nsearch", dist.Normal(jnp.log(20.0), 0.5))
    sigma_log_Nsearch = numpyro.sample("sigma_log_Nsearch", dist.HalfNormal(0.4))
    z_Nsearch = numpyro.sample("z_Nsearch", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_Nsearch_subj = mu_log_Nsearch + sigma_log_Nsearch * z_Nsearch
    N_search_subj = numpyro.deterministic("N_search_subj", jnp.exp(log_Nsearch_subj))

    # -----------------------------
    # Shared subjective expected interior path length
    # -----------------------------
    L_search = numpyro.sample("L_search", dist.Gamma(12.0, 0.5))  # mean 24

    # -----------------------------
    # Subject baseline + hard-condition shift in log lambda
    # This keeps your original idea that condition affects search cost
    # through lambda_c * N_search
    # -----------------------------
    mu_log_lambda = numpyro.sample("mu_log_lambda", dist.Normal(0.0, 0.5))
    sigma_log_lambda = numpyro.sample("sigma_log_lambda", dist.HalfNormal(0.3))
    z_lambda_base = numpyro.sample("z_lambda_base", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_lambda_base_subj = mu_log_lambda + sigma_log_lambda * z_lambda_base

    mu_delta_lambda = numpyro.sample("mu_delta_lambda", dist.Normal(0.0, 0.4))
    sigma_delta_lambda = numpyro.sample("sigma_delta_lambda", dist.HalfNormal(0.25))
    z_delta_lambda = numpyro.sample("z_delta_lambda", dist.Normal(0.0, 1.0).expand([n_subj]))
    delta_lambda_subj = mu_delta_lambda + sigma_delta_lambda * z_delta_lambda

    # -----------------------------
    # Trial-level indexing
    # -----------------------------
    beta_t = beta_subj[subj_idx]
    kappa_t = kappa_subj[subj_idx]
    rho_t = rho_subj[subj_idx]
    delta_reuse_t = delta_reuse_subj[subj_idx] * (cond_idx == 1)
    N_search_t = N_search_subj[subj_idx]

    log_lambda_t = log_lambda_base_subj[subj_idx] + delta_lambda_subj[subj_idx] * (cond_idx == 1)
    lambda_t = numpyro.deterministic("lambda_t", jnp.exp(log_lambda_t))

    # -----------------------------
    # Reuse utilities
    # Base value if open; blocked penalty if closed
    # Then add subject reuse bias and hard-condition reuse shift
    # -----------------------------
    U_left_base = jnp.where(
        cache_valid[:, 0] == 1,
        L_budget - rim_path_length[:, 0],
        -kappa_t
    )

    U_right_base = jnp.where(
        cache_valid[:, 1] == 1,
        L_budget - rim_path_length[:, 1],
        -kappa_t
    )

    U_left = U_left_base + rho_t + delta_reuse_t
    U_right = U_right_base + rho_t + delta_reuse_t

    # -----------------------------
    # Search utility
    # Keep simple: expected path gain minus condition-weighted inspection burden
    # -----------------------------
    U_search = (L_budget - L_search) - lambda_t * N_search_t

    utilities = jnp.stack([U_search, U_left, U_right], axis=1)
    logits = beta_t[:, None] * utilities

    numpyro.deterministic("utilities", utilities)
    numpyro.deterministic("logits", logits)
    numpyro.deterministic("obs_probs", jax.nn.softmax(logits, axis=1))

    with numpyro.plate("trials", T):
        numpyro.sample("obs", dist.Categorical(logits=logits), obs=obs_solution_slot)
```

```python
import jax
import jax.numpy as jnp
import numpyro
import numpyro.distributions as dist


def colored_maze_utility_hierarchical_v5(
    subj_idx,              # shape [T], integer-coded 0..S-1
    cond_idx,              # shape [T], integer-coded 0..C-1
    cache_valid,           # shape [T, 2], 1 if rim path valid/open, 0 otherwise
    rim_path_length,       # shape [T, 2], length of left/right rim path
    EU_search,             # kept for API compatibility; not used in this version
    EN_search,             # kept for API compatibility; not used in this version
    obs_solution_slot=None,# shape [T], 0=search, 1=left, 2=right
    n_subj=None,
    n_cond=None,
    L_budget=70.0,
):
    """
    Hierarchical static softmax model with:
      - subject-level inverse temperature beta_s
      - subject-level blocked-path aversion kappa_s
      - subject-level subjective search burden N_search_s
      - shared subjective expected interior path length L_search
      - subject-level baseline reuse bias rho_s
      - subject-level hard-condition reuse bonus delta_reuse_s
      - subject baseline + hard-condition shift in log lambda

    Action coding:
        0 = search
        1 = reuse-left
        2 = reuse-right

    Notes:
        EU_search and EN_search are accepted only to preserve the same
        function signature as earlier models. They are not used here.
    """
    if n_subj is None:
        raise ValueError("n_subj must be provided as a Python int.")
    if n_cond is None:
        raise ValueError("n_cond must be provided as a Python int.")
    if n_cond != 2:
        raise ValueError("This version assumes exactly 2 conditions.")

    T = subj_idx.shape[0]
    if cache_valid.shape != rim_path_length.shape:
        raise ValueError("cache_valid and rim_path_length must have same shape [T, 2].")
    if cache_valid.shape[1] != 2:
        raise ValueError("Expected exactly 2 rim paths: left and right.")

    # -----------------------------
    # Subject-level inverse temperature
    # -----------------------------
    mu_log_beta = numpyro.sample("mu_log_beta", dist.Normal(-1.0, 0.5))
    sigma_log_beta = numpyro.sample("sigma_log_beta", dist.HalfNormal(0.4))
    z_beta = numpyro.sample("z_beta", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_beta_subj = mu_log_beta + sigma_log_beta * z_beta
    beta_subj = numpyro.deterministic("beta_subj", jnp.exp(log_beta_subj))

    # -----------------------------
    # Subject-level blocked-path aversion
    # -----------------------------
    mu_log_kappa = numpyro.sample("mu_log_kappa", dist.Normal(2.5, 0.5))
    sigma_log_kappa = numpyro.sample("sigma_log_kappa", dist.HalfNormal(0.4))
    z_kappa = numpyro.sample("z_kappa", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_kappa_subj = mu_log_kappa + sigma_log_kappa * z_kappa
    kappa_subj = numpyro.deterministic("kappa_subj", jnp.exp(log_kappa_subj))

    # -----------------------------
    # Subject-level baseline reuse bias
    # Positive rho means cached reuse is more attractive overall
    # -----------------------------
    mu_rho = numpyro.sample("mu_rho", dist.Normal(0.0, 10.0))
    sigma_rho = numpyro.sample("sigma_rho", dist.HalfNormal(5.0))
    z_rho = numpyro.sample("z_rho", dist.Normal(0.0, 1.0).expand([n_subj]))
    rho_subj = numpyro.deterministic("rho_subj", mu_rho + sigma_rho * z_rho)

    # -----------------------------
    # Subject-level hard-condition reuse bonus
    # Only applies when cond_idx == 1
    # Positive delta means more reuse in hard condition
    # -----------------------------
    mu_delta_reuse = numpyro.sample("mu_delta_reuse", dist.Normal(0.0, 10.0))
    sigma_delta_reuse = numpyro.sample("sigma_delta_reuse", dist.HalfNormal(5.0))
    z_delta_reuse = numpyro.sample("z_delta_reuse", dist.Normal(0.0, 1.0).expand([n_subj]))
    delta_reuse_subj = numpyro.deterministic(
        "delta_reuse_subj",
        mu_delta_reuse + sigma_delta_reuse * z_delta_reuse
    )

    # -----------------------------
    # Subjective search burden: N_search_s
    # -----------------------------
    mu_log_Nsearch = numpyro.sample("mu_log_Nsearch", dist.Normal(jnp.log(20.0), 0.5))
    sigma_log_Nsearch = numpyro.sample("sigma_log_Nsearch", dist.HalfNormal(0.4))
    z_Nsearch = numpyro.sample("z_Nsearch", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_Nsearch_subj = mu_log_Nsearch + sigma_log_Nsearch * z_Nsearch
    N_search_subj = numpyro.deterministic("N_search_subj", jnp.exp(log_Nsearch_subj))

    # -----------------------------
    # Shared subjective expected interior path length
    # -----------------------------
#     L_search = numpyro.sample("L_search", dist.Gamma(12.0, 0.5))  # mean 24
    mu_log_Lsearch = numpyro.sample("mu_log_Lsearch", dist.Normal(jnp.log(20.0), 0.4))
    sigma_log_Lsearch = numpyro.sample("sigma_log_Lsearch", dist.HalfNormal(0.3))
    z_Lsearch = numpyro.sample("z_Lsearch", dist.Normal(0, 1).expand([n_subj]))
    log_Lsearch_subj = mu_log_Lsearch + sigma_log_Lsearch * z_Lsearch
    L_search_subj = numpyro.deterministic("L_search_subj", jnp.exp(log_Lsearch_subj))


    # -----------------------------
    # Subject baseline + hard-condition shift in log lambda
    # This keeps your original idea that condition affects search cost
    # through lambda_c * N_search
    # -----------------------------
    mu_log_lambda = numpyro.sample("mu_log_lambda", dist.Normal(0.0, 0.5))
    sigma_log_lambda = numpyro.sample("sigma_log_lambda", dist.HalfNormal(0.3))
    z_lambda_base = numpyro.sample("z_lambda_base", dist.Normal(0.0, 1.0).expand([n_subj]))
    log_lambda_base_subj = mu_log_lambda + sigma_log_lambda * z_lambda_base

    mu_delta_lambda = numpyro.sample("mu_delta_lambda", dist.Normal(0.0, 0.4))
    sigma_delta_lambda = numpyro.sample("sigma_delta_lambda", dist.HalfNormal(0.25))
    z_delta_lambda = numpyro.sample("z_delta_lambda", dist.Normal(0.0, 1.0).expand([n_subj]))
    delta_lambda_subj = mu_delta_lambda + sigma_delta_lambda * z_delta_lambda

    # -----------------------------
    # Trial-level indexing
    # -----------------------------
    beta_t = beta_subj[subj_idx]
    kappa_t = kappa_subj[subj_idx]
    rho_t = rho_subj[subj_idx]
    delta_reuse_t = delta_reuse_subj[subj_idx] * (cond_idx == 1)
    N_search_t = N_search_subj[subj_idx]
    L_search_t = L_search_subj[subj_idx]

    log_lambda_t = log_lambda_base_subj[subj_idx] + delta_lambda_subj[subj_idx] * (cond_idx == 1)
    lambda_t = numpyro.deterministic("lambda_t", jnp.exp(log_lambda_t))

    # -----------------------------
    # Reuse utilities
    # Base value if open; blocked penalty if closed
    # Then add subject reuse bias and hard-condition reuse shift
    # -----------------------------
    U_left_base = jnp.where(
        cache_valid[:, 0] == 1,
        L_budget - rim_path_length[:, 0],
        -kappa_t
    )

    U_right_base = jnp.where(
        cache_valid[:, 1] == 1,
        L_budget - rim_path_length[:, 1],
        -kappa_t
    )

    U_left = U_left_base + rho_t + delta_reuse_t
    U_right = U_right_base + rho_t + delta_reuse_t

    # -----------------------------
    # Search utility
    # Keep simple: expected path gain minus condition-weighted inspection burden
    # -----------------------------
    U_search = (L_budget - L_search_t) - lambda_t * N_search_t

    utilities = jnp.stack([U_search, U_left, U_right], axis=1)
    logits = beta_t[:, None] * utilities

    numpyro.deterministic("utilities", utilities)
    numpyro.deterministic("logits", logits)
    numpyro.deterministic("obs_probs", jax.nn.softmax(logits, axis=1))

    with numpyro.plate("trials", T):
        numpyro.sample("obs", dist.Categorical(logits=logits), obs=obs_solution_slot)
```

```python
import numpy as np
import pandas as pd
import jax.numpy as jnp
import jax.random as random
from numpyro.infer import Predictive


def posterior_predictive_compare_by_condition(
    posterior_samples,
    model_fn,
    inputs,
    df_subject,
    cond_col="cond",
    trial_col="trial",
    obs_col="obs_solution_slot",
    n_ppc_draws=None,
):
#     posterior_samples = mcmc.get_samples()

    if n_ppc_draws is not None:
        n_total = len(next(iter(posterior_samples.values())))
        idx = np.linspace(0, n_total - 1, n_ppc_draws).astype(int)
        posterior_samples = {k: np.asarray(v)[idx] for k, v in posterior_samples.items()}

    predictive = Predictive(
        model_fn,
        posterior_samples=posterior_samples,
        return_sites=["obs", "obs_probs", "utilities"],
    )

    ppc = predictive(
        random.PRNGKey(123),
        subj_idx=jnp.array(inputs["subj_idx"]),
        cond_idx=jnp.array(inputs["cond_idx"]),
        cache_valid=jnp.array(inputs["cache_valid"]),
        rim_path_length=jnp.array(inputs["rim_path_length"]),
        EU_search=jnp.array(inputs["EU_search"]),
        EN_search=jnp.array(inputs["EN_search"]),
        obs_solution_slot=None,
        n_subj=int(np.max(inputs["subj_idx"])) + 1,
        n_cond=int(np.max(inputs["cond_idx"])) + 1,
    )

    sampled_labels = np.asarray(ppc["obs"])              # [n_draws, T]
    obs_probs_post = np.asarray(ppc["obs_probs"])        # [n_draws, T, 3]
    utilities_post = np.asarray(ppc["utilities"])        # [n_draws, T, 3] or [T, 3]

    obs_probs_mean = obs_probs_post.mean(axis=0)         # [T, 3]

    sort_by = inputs.get("metadata", {}).get("sort_by", ("subj_idx","block", "trial"))
    df_aug = df_subject.sort_values(list(sort_by)).reset_index(drop=True).copy()

    T = len(inputs["cond_idx"])
    if len(df_aug) != T:
        raise ValueError(f"Dataframe has {len(df_aug)} rows, but model inputs have {T} trials.")

    df_aug["pp_mean_p_search"] = obs_probs_mean[:, 0]
    df_aug["pp_mean_p_reuse_left"] = obs_probs_mean[:, 1]
    df_aug["pp_mean_p_reuse_right"] = obs_probs_mean[:, 2]
    df_aug["pp_mean_p_reuse_total"] = obs_probs_mean[:, 1] + obs_probs_mean[:, 2]

    # Optional trialwise mean utilities
    if utilities_post.ndim == 3:
        utilities_mean = utilities_post.mean(axis=0)
    else:
        utilities_mean = utilities_post

    df_aug["pp_mean_u_search"] = utilities_mean[:, 0]
    df_aug["pp_mean_u_reuse_left"] = utilities_mean[:, 1]
    df_aug["pp_mean_u_reuse_right"] = utilities_mean[:, 2]

    # Actual human summaries by condition
    actual_rows = []
    for cond_val, d in df_aug.groupby(cond_col, sort=True):
        y = d[obs_col].to_numpy()
        n = len(y)
        n_search = np.sum(y == 0)
        n_reuse = np.sum(np.isin(y, [1, 2]))
        n_left = np.sum(y == 1)
        n_right = np.sum(y == 2)

        actual_rows.append({
            cond_col: cond_val,
            "n_trials": n,
            "p_search": n_search / n,
            "p_reuse": n_reuse / n,
            "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
            "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
            "n_search": n_search,
            "n_reuse": n_reuse,
            "n_left": n_left,
            "n_right": n_right,
        })

    actual_by_cond = (
        pd.DataFrame(actual_rows)
        .sort_values(cond_col)
        .reset_index(drop=True)
    )

    # PPC summaries by draw and condition
    cond_values = df_aug[cond_col].to_numpy()
    unique_conds = np.sort(df_aug[cond_col].unique())

    ppc_rows = []
    for s in range(sampled_labels.shape[0]):
        ys = sampled_labels[s]
        for cond_val in unique_conds:
            mask = (cond_values == cond_val)
            yc = ys[mask]
            n = len(yc)

            n_search = np.sum(yc == 0)
            n_reuse = np.sum(np.isin(yc, [1, 2]))
            n_left = np.sum(yc == 1)
            n_right = np.sum(yc == 2)

            ppc_rows.append({
                "draw": s,
                cond_col: cond_val,
                "n_trials": n,
                "p_search": n_search / n,
                "p_reuse": n_reuse / n,
                "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
                "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
                "n_search": n_search,
                "n_reuse": n_reuse,
                "n_left": n_left,
                "n_right": n_right,
            })

    ppc_by_cond_draw = pd.DataFrame(ppc_rows)

    metrics = ["p_search", "p_reuse", "p_left_given_reuse", "p_right_given_reuse"]
    comp_rows = []

    for cond_val in unique_conds:
        actual_row = actual_by_cond.loc[actual_by_cond[cond_col] == cond_val].iloc[0]
        ppc_sub = ppc_by_cond_draw.loc[ppc_by_cond_draw[cond_col] == cond_val]

        for metric in metrics:
            vals = ppc_sub[metric].dropna().to_numpy()
            comp_rows.append({
                cond_col: cond_val,
                "metric": metric,
                "actual": actual_row[metric],
                "ppc_mean": np.mean(vals),
                "ppc_sd": np.std(vals),
                "ppc_q05": np.quantile(vals, 0.05),
                "ppc_q50": np.quantile(vals, 0.50),
                "ppc_q95": np.quantile(vals, 0.95),
            })

    comparison_by_cond = (
        pd.DataFrame(comp_rows)
        .sort_values([cond_col, "metric"])
        .reset_index(drop=True)
    )

    return {
        "sampled_labels": sampled_labels,
        "obs_probs_post": obs_probs_post,
        "utilities_post": utilities_post,
        "df_aug": df_aug,
        "actual_by_cond": actual_by_cond,
        "ppc_by_cond_draw": ppc_by_cond_draw,
        "comparison_by_cond": comparison_by_cond,
    }
```

```python
import numpy as np
import pandas as pd


def summarize_ppc_by_subject_condition(
    ppc_out,
    subj_col="subj_idx",
    cond_col="cond",
    obs_col="obs_solution_slot",
):
    """
    Summarize actual and posterior predictive behavior by subject x condition.

    Returns
    -------
    dict with:
        actual_by_subj_cond : DataFrame
        ppc_by_subj_cond_draw : DataFrame
        comparison_by_subj_cond : DataFrame
    """
    df_aug = ppc_out["df_aug"].copy()
    sampled_labels = np.asarray(ppc_out["sampled_labels"])   # [n_draws, T]

    if sampled_labels.ndim != 2:
        raise ValueError("Expected sampled_labels to have shape [n_draws, T].")

    T = len(df_aug)
    if sampled_labels.shape[1] != T:
        raise ValueError(f"sampled_labels has T={sampled_labels.shape[1]} but df_aug has {T} rows.")

    # -------------------------
    # Actual summaries
    # -------------------------
    actual_rows = []
    grouped = df_aug.groupby([subj_col, cond_col], sort=True)

    for (subj_val, cond_val), d in grouped:
        y = d[obs_col].to_numpy()
        n = len(y)

        n_search = np.sum(y == 0)
        n_reuse = np.sum(np.isin(y, [1, 2]))
        n_left = np.sum(y == 1)
        n_right = np.sum(y == 2)

        actual_rows.append({
            subj_col: subj_val,
            cond_col: cond_val,
            "n_trials": n,
            "p_search": n_search / n,
            "p_reuse": n_reuse / n,
            "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
            "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
            "n_search": n_search,
            "n_reuse": n_reuse,
            "n_left": n_left,
            "n_right": n_right,
        })

    actual_by_subj_cond = (
        pd.DataFrame(actual_rows)
        .sort_values([subj_col, cond_col])
        .reset_index(drop=True)
    )

    # -------------------------
    # PPC summaries by draw
    # -------------------------
    subj_vals = df_aug[subj_col].to_numpy()
    cond_vals = df_aug[cond_col].to_numpy()
    unique_pairs = (
        df_aug[[subj_col, cond_col]]
        .drop_duplicates()
        .sort_values([subj_col, cond_col])
        .to_records(index=False)
    )

    ppc_rows = []
    for draw in range(sampled_labels.shape[0]):
        ys = sampled_labels[draw]

        for subj_val, cond_val in unique_pairs:
            mask = (subj_vals == subj_val) & (cond_vals == cond_val)
            yc = ys[mask]
            n = len(yc)

            n_search = np.sum(yc == 0)
            n_reuse = np.sum(np.isin(yc, [1, 2]))
            n_left = np.sum(yc == 1)
            n_right = np.sum(yc == 2)

            ppc_rows.append({
                "draw": draw,
                subj_col: subj_val,
                cond_col: cond_val,
                "n_trials": n,
                "p_search": n_search / n,
                "p_reuse": n_reuse / n,
                "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
                "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
                "n_search": n_search,
                "n_reuse": n_reuse,
                "n_left": n_left,
                "n_right": n_right,
            })

    ppc_by_subj_cond_draw = pd.DataFrame(ppc_rows)

    # -------------------------
    # Comparison table:
    # actual vs posterior predictive mean per subject x condition
    # -------------------------
    metrics = ["p_search", "p_reuse", "p_left_given_reuse", "p_right_given_reuse"]
    comp_rows = []

    for (subj_val, cond_val), arow in actual_by_subj_cond.groupby([subj_col, cond_col], sort=True):
        arow = arow.iloc[0]
        ppc_sub = ppc_by_subj_cond_draw[
            (ppc_by_subj_cond_draw[subj_col] == subj_val) &
            (ppc_by_subj_cond_draw[cond_col] == cond_val)
        ]

        for metric in metrics:
            vals = ppc_sub[metric].dropna().to_numpy()
            comp_rows.append({
                subj_col: subj_val,
                cond_col: cond_val,
                "metric": metric,
                "actual": arow[metric],
                "ppc_mean": np.mean(vals) if len(vals) else np.nan,
                "ppc_sd": np.std(vals) if len(vals) else np.nan,
                "ppc_q05": np.quantile(vals, 0.05) if len(vals) else np.nan,
                "ppc_q50": np.quantile(vals, 0.50) if len(vals) else np.nan,
                "ppc_q95": np.quantile(vals, 0.95) if len(vals) else np.nan,
            })

    comparison_by_subj_cond = (
        pd.DataFrame(comp_rows)
        .sort_values([subj_col, cond_col, "metric"])
        .reset_index(drop=True)
    )

    return {
        "actual_by_subj_cond": actual_by_subj_cond,
        "ppc_by_subj_cond_draw": ppc_by_subj_cond_draw,
        "comparison_by_subj_cond": comparison_by_subj_cond,
    }



import numpy as np
import pandas as pd


def summarize_ppc_by_subject_condition(
    ppc_out,
    subj_col="subj_idx",
    cond_col="cond",
    obs_col="obs_solution_slot",
):
    """
    Summarize actual and posterior predictive behavior by subject x condition.

    Returns
    -------
    dict with:
        actual_by_subj_cond : DataFrame
        ppc_by_subj_cond_draw : DataFrame
        comparison_by_subj_cond : DataFrame
    """
    df_aug = ppc_out["df_aug"].copy()
    sampled_labels = np.asarray(ppc_out["sampled_labels"])   # [n_draws, T]

    if sampled_labels.ndim != 2:
        raise ValueError("Expected sampled_labels to have shape [n_draws, T].")

    T = len(df_aug)
    if sampled_labels.shape[1] != T:
        raise ValueError(f"sampled_labels has T={sampled_labels.shape[1]} but df_aug has {T} rows.")

    # -------------------------
    # Actual summaries
    # -------------------------
    actual_rows = []
    grouped = df_aug.groupby([subj_col, cond_col], sort=True)

    for (subj_val, cond_val), d in grouped:
        y = d[obs_col].to_numpy()
        n = len(y)

        n_search = np.sum(y == 0)
        n_reuse = np.sum(np.isin(y, [1, 2]))
        n_left = np.sum(y == 1)
        n_right = np.sum(y == 2)

        actual_rows.append({
            subj_col: subj_val,
            cond_col: cond_val,
            "n_trials": n,
            "p_search": n_search / n,
            "p_reuse": n_reuse / n,
            "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
            "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
            "n_search": n_search,
            "n_reuse": n_reuse,
            "n_left": n_left,
            "n_right": n_right,
        })

    actual_by_subj_cond = (
        pd.DataFrame(actual_rows)
        .sort_values([subj_col, cond_col])
        .reset_index(drop=True)
    )

    # -------------------------
    # PPC summaries by draw
    # -------------------------
    subj_vals = df_aug[subj_col].to_numpy()
    cond_vals = df_aug[cond_col].to_numpy()
    unique_pairs = (
        df_aug[[subj_col, cond_col]]
        .drop_duplicates()
        .sort_values([subj_col, cond_col])
        .to_records(index=False)
    )

    ppc_rows = []
    for draw in range(sampled_labels.shape[0]):
        ys = sampled_labels[draw]

        for subj_val, cond_val in unique_pairs:
            mask = (subj_vals == subj_val) & (cond_vals == cond_val)
            yc = ys[mask]
            n = len(yc)

            n_search = np.sum(yc == 0)
            n_reuse = np.sum(np.isin(yc, [1, 2]))
            n_left = np.sum(yc == 1)
            n_right = np.sum(yc == 2)

            ppc_rows.append({
                "draw": draw,
                subj_col: subj_val,
                cond_col: cond_val,
                "n_trials": n,
                "p_search": n_search / n,
                "p_reuse": n_reuse / n,
                "p_left_given_reuse": n_left / n_reuse if n_reuse > 0 else np.nan,
                "p_right_given_reuse": n_right / n_reuse if n_reuse > 0 else np.nan,
                "n_search": n_search,
                "n_reuse": n_reuse,
                "n_left": n_left,
                "n_right": n_right,
            })

    ppc_by_subj_cond_draw = pd.DataFrame(ppc_rows)

    # -------------------------
    # Comparison table:
    # actual vs posterior predictive mean per subject x condition
    # -------------------------
    metrics = ["p_search", "p_reuse", "p_left_given_reuse", "p_right_given_reuse"]
    comp_rows = []

    for (subj_val, cond_val), arow in actual_by_subj_cond.groupby([subj_col, cond_col], sort=True):
        arow = arow.iloc[0]
        ppc_sub = ppc_by_subj_cond_draw[
            (ppc_by_subj_cond_draw[subj_col] == subj_val) &
            (ppc_by_subj_cond_draw[cond_col] == cond_val)
        ]

        for metric in metrics:
            vals = ppc_sub[metric].dropna().to_numpy()
            comp_rows.append({
                subj_col: subj_val,
                cond_col: cond_val,
                "metric": metric,
                "actual": arow[metric],
                "ppc_mean": np.mean(vals) if len(vals) else np.nan,
                "ppc_sd": np.std(vals) if len(vals) else np.nan,
                "ppc_q05": np.quantile(vals, 0.05) if len(vals) else np.nan,
                "ppc_q50": np.quantile(vals, 0.50) if len(vals) else np.nan,
                "ppc_q95": np.quantile(vals, 0.95) if len(vals) else np.nan,
            })

    comparison_by_subj_cond = (
        pd.DataFrame(comp_rows)
        .sort_values([subj_col, cond_col, "metric"])
        .reset_index(drop=True)
    )

    return {
        "actual_by_subj_cond": actual_by_subj_cond,
        "ppc_by_subj_cond_draw": ppc_by_subj_cond_draw,
        "comparison_by_subj_cond": comparison_by_subj_cond,
    }

import matplotlib.pyplot as plt
import numpy as np


def plot_ppc_subject_scatter_by_condition(
    subj_ppc_out,
    cond_col="cond",
    subj_col="subj_idx",
    add_errorbars=False,
):
    """
    Scatterplot actual vs posterior predictive mean, one point per subject,
    separately by condition and metric.
    """
    comp = subj_ppc_out["comparison_by_subj_cond"].copy()

    metrics = [
        ("p_search", "P(search)"),
        ("p_reuse", "P(reuse)"),
        ("p_left_given_reuse", "P(left | reuse)"),
        ("p_right_given_reuse", "P(right | reuse)"),
    ]

    cond_vals = np.sort(comp[cond_col].unique())
    nrows = len(cond_vals)
    ncols = len(metrics)

    fig, axes = plt.subplots(nrows, ncols, figsize=(4 * ncols, 4 * nrows), squeeze=False)

    for i, cond_val in enumerate(cond_vals):
        for j, (metric, title) in enumerate(metrics):
            ax = axes[i, j]
            d = comp[(comp[cond_col] == cond_val) & (comp["metric"] == metric)].copy()
            d = d.dropna(subset=["actual", "ppc_mean"])

            x = d["actual"].to_numpy()
            y = d["ppc_mean"].to_numpy()

            if len(x) == 0:
                ax.set_visible(False)
                continue

            ax.scatter(x, y, alpha=0.2)

            if add_errorbars:
                yerr_low = y - d["ppc_q05"].to_numpy()
                yerr_high = d["ppc_q95"].to_numpy() - y
                ax.errorbar(x, y, yerr=[yerr_low, yerr_high], fmt="none", alpha=0.5)

            lo = np.nanmin(np.concatenate([x, y]))
            hi = np.nanmax(np.concatenate([x, y]))
            pad = 0.05 * (hi - lo if hi > lo else 1.0)

            ax.plot([lo - pad, hi + pad], [lo - pad, hi + pad], linestyle="--")
            ax.set_xlim(lo - pad, hi + pad)
            ax.set_ylim(lo - pad, hi + pad)

            ax.set_title(f"{title}\ncond={cond_val}")
            ax.set_xlabel("Actual")
            ax.set_ylabel("Posterior predictive mean")

    plt.tight_layout()
    plt.show()

def summarize_ppc_subject_condition_differences(
    subj_ppc_out,
    subj_col="subj_idx",
    cond_col="cond",
):
    """
    Compute subject-level condition differences:
        higher-coded condition - lower-coded condition

    Returns actual and PPC mean differences by subject and metric.
    """
    actual_df = subj_ppc_out["actual_by_subj_cond"].copy()
    ppc_draw_df = subj_ppc_out["ppc_by_subj_cond_draw"].copy()

    cond_vals = np.sort(actual_df[cond_col].unique())
    if len(cond_vals) != 2:
        raise ValueError(f"Expected exactly 2 conditions, got {cond_vals}")

    c0, c1 = cond_vals[0], cond_vals[1]
    metrics = ["p_search", "p_reuse", "p_left_given_reuse", "p_right_given_reuse"]

    # Actual diffs by subject
    actual_rows = []
    for subj_val, dsub in actual_df.groupby(subj_col, sort=True):
        r0 = dsub[dsub[cond_col] == c0]
        r1 = dsub[dsub[cond_col] == c1]
        if len(r0) != 1 or len(r1) != 1:
            continue
        r0 = r0.iloc[0]
        r1 = r1.iloc[0]

        for metric in metrics:
            actual_rows.append({
                subj_col: subj_val,
                "metric": metric,
                "actual_diff": r1[metric] - r0[metric],
                "cond_low": c0,
                "cond_high": c1,
            })

    actual_diff_by_subj = pd.DataFrame(actual_rows)

    # PPC diffs by subject and draw
    ppc_rows = []
    for (draw, subj_val), dsub in ppc_draw_df.groupby(["draw", subj_col], sort=True):
        r0 = dsub[dsub[cond_col] == c0]
        r1 = dsub[dsub[cond_col] == c1]
        if len(r0) != 1 or len(r1) != 1:
            continue
        r0 = r0.iloc[0]
        r1 = r1.iloc[0]

        for metric in metrics:
            ppc_rows.append({
                "draw": draw,
                subj_col: subj_val,
                "metric": metric,
                "ppc_diff": r1[metric] - r0[metric],
                "cond_low": c0,
                "cond_high": c1,
            })

    ppc_diff_by_subj_draw = pd.DataFrame(ppc_rows)

    # Merge summary
    comp_rows = []
    for (subj_val, metric), arow in actual_diff_by_subj.groupby([subj_col, "metric"], sort=True):
        arow = arow.iloc[0]
        vals = ppc_diff_by_subj_draw[
            (ppc_diff_by_subj_draw[subj_col] == subj_val) &
            (ppc_diff_by_subj_draw["metric"] == metric)
        ]["ppc_diff"].dropna().to_numpy()

        comp_rows.append({
            subj_col: subj_val,
            "metric": metric,
            "actual_diff": arow["actual_diff"],
            "ppc_mean_diff": np.mean(vals) if len(vals) else np.nan,
            "ppc_sd_diff": np.std(vals) if len(vals) else np.nan,
            "ppc_q05_diff": np.quantile(vals, 0.05) if len(vals) else np.nan,
            "ppc_q50_diff": np.quantile(vals, 0.50) if len(vals) else np.nan,
            "ppc_q95_diff": np.quantile(vals, 0.95) if len(vals) else np.nan,
        })

    comparison_diff_by_subj = pd.DataFrame(comp_rows).sort_values([subj_col, "metric"]).reset_index(drop=True)

    return {
        "actual_diff_by_subj": actual_diff_by_subj,
        "ppc_diff_by_subj_draw": ppc_diff_by_subj_draw,
        "comparison_diff_by_subj": comparison_diff_by_subj,
    }


def plot_ppc_subject_condition_difference_scatter(
    diff_out,
    subj_col="subj_idx",
    add_errorbars=False,
):
    """
    Scatterplot subject-level actual condition difference vs posterior predictive mean condition difference.
    """
    comp = diff_out["comparison_diff_by_subj"].copy()

    metrics = [
        ("p_search", "Δ P(search)"),
        ("p_reuse", "Δ P(reuse)"),
        ("p_left_given_reuse", "Δ P(left | reuse)"),
        ("p_right_given_reuse", "Δ P(right | reuse)"),
    ]

    fig, axes = plt.subplots(1, len(metrics), figsize=(4 * len(metrics), 4), squeeze=False)
    axes = axes[0]

    for ax, (metric, title) in zip(axes, metrics):
        d = comp[comp["metric"] == metric].dropna(subset=["actual_diff", "ppc_mean_diff"]).copy()

        x = d["actual_diff"].to_numpy()
        y = d["ppc_mean_diff"].to_numpy()

        if len(x) == 0:
            ax.set_visible(False)
            continue

        ax.scatter(x, y, alpha=0.2)

        if add_errorbars:
            yerr_low = y - d["ppc_q05_diff"].to_numpy()
            yerr_high = d["ppc_q95_diff"].to_numpy() - y
            ax.errorbar(x, y, yerr=[yerr_low, yerr_high], fmt="none", alpha=0.5)

        lo = np.nanmin(np.concatenate([x, y]))
        hi = np.nanmax(np.concatenate([x, y]))
        pad = 0.05 * (hi - lo if hi > lo else 1.0)

        ax.plot([lo - pad, hi + pad], [lo - pad, hi + pad], linestyle="--")
        ax.set_xlim(lo - pad, hi + pad)
        ax.set_ylim(lo - pad, hi + pad)

        ax.set_title(title)
        ax.set_xlabel("Actual condition difference")
        ax.set_ylabel("Posterior predictive mean difference")

    plt.tight_layout()
    plt.show()
    
    
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def _infer_open_slot_from_cache_valid(cache_valid_entry):
    """
    Convert cache_valid like [1, 0] / [0, 1] into open_slot coded as 1(left) / 2(right).
    Returns np.nan if not exactly one slot is open.
    """
    arr = np.asarray(cache_valid_entry).astype(int).ravel()
    if arr.shape[0] != 2:
        return np.nan
    if np.array_equal(arr, [1, 0]):
        return 1
    if np.array_equal(arr, [0, 1]):
        return 2
    return np.nan


def summarize_open_rim_choice_ppc_by_subject_condition(
    ppc_obs,
    df_subject,
    subj_col="subj_idx",
    cond_col="cond",
    obs_col="obs_solution_slot",
    sort_by=("subj_idx", "block", "trial"),
    cache_valid_col="cache_valid",
    open_slot_col=None,
):
    """
    Summarize the statistic:

    'Across trials where exactly one rim path is open and the subject submitted a rim path,
     what is the probability that the chosen rim path was the open one?'

    Parameters
    ----------
    ppc_obs : array-like, shape [n_draws, T]
        Posterior predictive sampled labels, coded 0=search, 1=left, 2=right.
    df_subject : pd.DataFrame
        Trialwise dataframe aligned to the modeled trials.
    subj_col, cond_col, obs_col : str
        Column names in df_subject.
    sort_by : tuple/list
        Sorting used to align df_subject to the model input order.
    cache_valid_col : str
        Column containing 2-element validity vectors like [1,0] or [0,1].
    open_slot_col : str or None
        If provided, use this column directly as the open rim slot (1 or 2).
        If None, infer it from cache_valid_col.

    Returns
    -------
    dict with:
        df_aug
        actual_by_subj_cond
        ppc_by_subj_cond_draw
        comparison_by_subj_cond
    """
    ppc_obs = np.asarray(ppc_obs)
    if ppc_obs.ndim != 2:
        raise ValueError("ppc_obs must have shape [n_draws, T].")

    df_aug = df_subject.sort_values(list(sort_by)).reset_index(drop=True).copy()
    T = len(df_aug)

    if ppc_obs.shape[1] != T:
        raise ValueError(f"ppc_obs has T={ppc_obs.shape[1]} but df_subject has {T} rows after sorting.")

    # Determine which rim slot is open on each trial: 1=left, 2=right, nan otherwise
    if open_slot_col is not None:
        df_aug["open_slot"] = df_aug[open_slot_col]
    else:
        if cache_valid_col not in df_aug.columns:
            raise ValueError(
                f"Need either open_slot_col or a {cache_valid_col!r} column to infer which rim is open."
            )
        df_aug["open_slot"] = df_aug[cache_valid_col].apply(_infer_open_slot_from_cache_valid)

    # Eligible trials: exactly one rim open
    df_aug["one_rim_open"] = df_aug["open_slot"].isin([1, 2])

    # Actual: submitted a rim path
    df_aug["actual_submitted_rim"] = df_aug[obs_col].isin([1, 2])

    # Actual: chose the open rim path
    df_aug["actual_chose_open_rim"] = (
        df_aug["actual_submitted_rim"] &
        (df_aug[obs_col] == df_aug["open_slot"])
    )

    # ---------- Actual summaries ----------
    actual_rows = []
    for (subj_val, cond_val), d in df_aug.groupby([subj_col, cond_col], sort=True):
        elig = d["one_rim_open"] & d["actual_submitted_rim"]
        n_elig = int(np.sum(elig))
        n_correct = int(np.sum(d.loc[elig, "actual_chose_open_rim"]))

        actual_rows.append({
            subj_col: subj_val,
            cond_col: cond_val,
            "n_eligible": n_elig,
            "n_open_rim_chosen": n_correct,
            "p_open_rim_given_rim_submit": (n_correct / n_elig) if n_elig > 0 else np.nan,
        })

    actual_by_subj_cond = (
        pd.DataFrame(actual_rows)
        .sort_values([subj_col, cond_col])
        .reset_index(drop=True)
    )

    # ---------- PPC summaries by draw ----------
    open_slot = df_aug["open_slot"].to_numpy()
    one_rim_open = df_aug["one_rim_open"].to_numpy()
    subj_vals = df_aug[subj_col].to_numpy()
    cond_vals = df_aug[cond_col].to_numpy()

    unique_pairs = (
        df_aug[[subj_col, cond_col]]
        .drop_duplicates()
        .sort_values([subj_col, cond_col])
        .to_records(index=False)
    )

    ppc_rows = []
    for draw in range(ppc_obs.shape[0]):
        y = ppc_obs[draw]  # shape [T]

        submitted_rim = np.isin(y, [1, 2])
        chose_open_rim = submitted_rim & (y == open_slot)

        for subj_val, cond_val in unique_pairs:
            mask_pair = (subj_vals == subj_val) & (cond_vals == cond_val)
            elig = mask_pair & one_rim_open & submitted_rim

            n_elig = int(np.sum(elig))
            n_correct = int(np.sum(chose_open_rim & elig))

            ppc_rows.append({
                "draw": draw,
                subj_col: subj_val,
                cond_col: cond_val,
                "n_eligible": n_elig,
                "n_open_rim_chosen": n_correct,
                "p_open_rim_given_rim_submit": (n_correct / n_elig) if n_elig > 0 else np.nan,
            })

    ppc_by_subj_cond_draw = pd.DataFrame(ppc_rows)

    # ---------- Comparison table ----------
    comp_rows = []
    for (subj_val, cond_val), arow in actual_by_subj_cond.groupby([subj_col, cond_col], sort=True):
        arow = arow.iloc[0]
        vals = ppc_by_subj_cond_draw.loc[
            (ppc_by_subj_cond_draw[subj_col] == subj_val) &
            (ppc_by_subj_cond_draw[cond_col] == cond_val),
            "p_open_rim_given_rim_submit"
        ].dropna().to_numpy()

        comp_rows.append({
            subj_col: subj_val,
            cond_col: cond_val,
            "actual": arow["p_open_rim_given_rim_submit"],
            "ppc_mean": np.mean(vals) if len(vals) else np.nan,
            "ppc_sd": np.std(vals) if len(vals) else np.nan,
            "ppc_q05": np.quantile(vals, 0.05) if len(vals) else np.nan,
            "ppc_q50": np.quantile(vals, 0.50) if len(vals) else np.nan,
            "ppc_q95": np.quantile(vals, 0.95) if len(vals) else np.nan,
            "n_eligible_actual": arow["n_eligible"],
        })

    comparison_by_subj_cond = (
        pd.DataFrame(comp_rows)
        .sort_values([subj_col, cond_col])
        .reset_index(drop=True)
    )

    return {
        "df_aug": df_aug,
        "actual_by_subj_cond": actual_by_subj_cond,
        "ppc_by_subj_cond_draw": ppc_by_subj_cond_draw,
        "comparison_by_subj_cond": comparison_by_subj_cond,
    }


def plot_open_rim_choice_ppc_scatter(
    open_rim_ppc_out,
    cond_col="cond",
    subj_col="subj_idx",
    add_errorbars=False,
    min_eligible_trials=1,
):
    """
    Scatterplot of:
      x = actual human P(chose open rim | one rim open and submitted rim)
      y = posterior predictive mean of the same statistic

    Separate panel for each condition.
    Each point is one subject.
    """
    comp = open_rim_ppc_out["comparison_by_subj_cond"].copy()
    comp = comp[comp["n_eligible_actual"] >= min_eligible_trials].copy()

    cond_vals = np.sort(comp[cond_col].unique())
    fig, axes = plt.subplots(1, len(cond_vals), figsize=(3.5 * len(cond_vals), 4), squeeze=False)
    axes = axes[0]

    for ax, cond_val in zip(axes, cond_vals):
        d = comp[comp[cond_col] == cond_val].dropna(subset=["actual", "ppc_mean"]).copy()

        x = d["actual"].to_numpy()
        y = d["ppc_mean"].to_numpy()

        if len(x) == 0:
            ax.set_visible(False)
            continue

        ax.scatter(x, y, alpha=0.2)

        if add_errorbars:
            yerr_low = y - d["ppc_q05"].to_numpy()
            yerr_high = d["ppc_q95"].to_numpy() - y
            ax.errorbar(x, y, yerr=[yerr_low, yerr_high], fmt="none", alpha=0.5)

        lo = np.nanmin(np.concatenate([x, y, np.array([0.0, 1.0])]))
        hi = np.nanmax(np.concatenate([x, y, np.array([0.0, 1.0])]))
        pad = 0.05 * (hi - lo if hi > lo else 1.0)

        ax.plot([lo - pad, hi + pad], [lo - pad, hi + pad], linestyle="--")
        ax.set_xlim(lo - pad, hi + pad)
        ax.set_ylim(lo - pad, hi + pad)

        ax.set_title(f"cond={cond_val}")
        ax.set_xlabel("Actual")
        ax.set_ylabel("Posterior predictive mean")

    fig.suptitle("P(submit correct rim path | only one is open)")
    plt.tight_layout()
    plt.show()
```

## ELPD-LOO model comparison

```python
import numpy as np
import arviz as az
import jax.numpy as jnp
from numpyro.infer import log_likelihood


def compute_loo_for_colored_maze_model_from_samples(
    posterior_samples,
    model_fn,
    inputs,
):
    """
    Compute PSIS-LOO / ELPD-LOO for a fitted NumPyro model using a posterior_samples dict.

    Parameters
    ----------
    posterior_samples : dict
        Dictionary from mcmc.get_samples(group_by_chain=False).
        Each value should have shape (draws, ...) with no chain dimension.
    model_fn : callable
        The NumPyro model function.
    inputs : dict
        Model inputs.

    Returns
    -------
    loo_res : arviz.ELPDData
    idata : arviz.InferenceData
    """
    # Keep these exactly as (draws, ...)
    posterior_samples_unchained = {
        k: np.asarray(v)
        for k, v in posterior_samples.items()
    }

    # log_likelihood expects group_by_chain=False style samples by default
    log_lik = log_likelihood(
        model_fn,
        posterior_samples_unchained,
        subj_idx=jnp.array(inputs["subj_idx"]),
        cond_idx=jnp.array(inputs["cond_idx"]),
        cache_valid=jnp.array(inputs["cache_valid"]),
        rim_path_length=jnp.array(inputs["rim_path_length"]),
        EU_search=jnp.array(inputs["EU_search"]),
        EN_search=jnp.array(inputs["EN_search"]),
        obs_solution_slot=jnp.array(inputs["obs_solution_slot"]),
        n_subj=int(np.max(inputs["subj_idx"])) + 1,
        n_cond=int(np.max(inputs["cond_idx"])) + 1,
    )

    # Now add singleton chain dimension only for ArviZ
    posterior_samples_chained = {
        k: np.expand_dims(v, axis=0)   # (draws, ...) -> (1, draws, ...)
        for k, v in posterior_samples_unchained.items()
    }

    log_lik_chained = {
        "obs": np.expand_dims(np.asarray(log_lik["obs"]), axis=0)  # (draws, T) -> (1, draws, T)
    }

    idata = az.from_dict(
        posterior=posterior_samples_chained,
        log_likelihood=log_lik_chained,
    )

    loo_res = az.loo(idata, var_name="obs", pointwise=True)
    return loo_res, idata
```
